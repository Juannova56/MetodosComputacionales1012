```python
def numeros_defectivos(a): # determina si un numero es defectivo, es decir la suma de sus divisores es menor que el propio numero.
                           # entrada "a": numero a evaluar
  if a <=1:                # retorna bool: True or False
    return False
  suma_divisores=0
  for i in range(1,a):
    if a%i==0:
      suma_divisores+=i

  return suma_divisores<a
print(numeros_defectivos(10))
print(numeros_defectivos(6))
```

```python
def numeros_abundantes(a):
    """
    Determina si un número es abundante (suma de divisores propios > número)

    Args:
        a (int): Número a evaluar

    Returns:
        bool: True si es abundante, False si no
    """
    if a <= 1:
        return False

    suma_divisores = 0
    for i in range(1, a):
        if a % i == 0:
            suma_divisores += i

    return suma_divisores > a

# Ejemplo de uso:
print(numeros_abundantes(12))  # True (1+2+3+4+6=16 > 12)
print(numeros_abundantes(6))   # False (1+2+3=6 no es mayor)
print(numeros_abundantes(1))   # False
```

```python
def numeros_perfectos(a): #determina si un numero es perfecto (es igual a la suma de sus divisores)
                          # entrada= numero a evaluar
  if a<=1:                # salida booleano (true or false)
    return False
  suma_divisores=0
  for i in range(1,a):
    if a%i==0:
     suma_divisores +=i
  return suma_divisores==a
print(numeros_perfectos(6))
print(numeros_perfectos(28))
```

```python
def numeros_primos(a):
  """determina si un numero es primo
  entrada: a(int) numero a evaluar
  retorna: bool: True si es primo, false si no"""
  if a<=1:
    return False
  if a==2:
    return True
  if a%2==0:
    return False
  for i in range(3,int(a**0.5)+1,2):
     if a%i==0:
      return False
  return True
print(numeros_primos(7))
print(numeros_primos(15))
```

```python
def numeros_semiperfectos(a):
  """determina si un numero es semiperfecto
  entrada: a(int) Numero a evaluar
  """
  if a<=1:
     return False
  divisores=[i for i in range(1,a) if a%i==0]
  sumas_posibles={0}
  for j in divisores:
    nuevas_sumas=set()
    for k in  sumas_posibles:
      suma=j+k
      if suma==a:
        return True
      if suma < a:
        nuevas_sumas.add(suma)
    sumas_posibles.update(nuevas_sumas)
  return False
print(numeros_semiperfectos(12))
print(numeros_semiperfectos(20))
```

3) numeros de catalan y maximo comun divisor con recursiones

```python
def catalan(n):
    """
    Calcula el n-ésimo número de Catalan de forma recursiva

    Args:
        n (int): Índice del número de Catalan (n >= 0)

    Returns:
        int: El n-ésimo número de Catalan

    Raises:
        ValueError: Si n es negativo
    """

    if n == 0:
        return 1
    # Validación para números negativos
    elif n < 0:
        raise ValueError("n debe ser un entero no negativo")
    # Caso recursivo
    else:
        return sum(catalan(i) * catalan(n - 1 - i) for i in range(n))
# imprimir los primeros 50 números de catalan
for i in range(50):
    print(f"C_{i} = {catalan(i)}")

```

3) maximo comun divisor

```python
def g(m, n):
    """
    Calcula el máximo común divisor (MCD) de m y n usando el algoritmo de Euclides con recursión

    Args:
        m (int): Primer número entero
        n (int): Segundo número entero

    Returns:
        int: El MCD de m y n
    """
    # Caso base: cuando n es 0, el MCD es m
    if n == 0:
        return m
    # Paso recursivo: MCD(m, n) = MCD(n, m mod n)
    else:
        return g(n, m % n)

# Calcular e imprimir el MCD de 108 y 192
mcd = g(108, 192)
print(f"El máximo común divisor de 108 y 192 es: {mcd}")
```

4) probabilidad cuantica

```python
import math
import scipy.constants as const

def calcular_probabilidades_cuanticas(E, V, m):
    """
    Calcula las probabilidades de transmisión (T) y reflexión (R) para una partícula cuántica
    que encuentra un escalón de potencial.

    Args:
        E (float): Energía de la partícula en eV
        V (float): Altura del escalón de potencial en eV
        m (float): Masa de la partícula en kg

    Returns:
        tuple: (T, R) probabilidades de transmisión y reflexión
    """
    # convertir energías de eV a julios
    E_J = E * const.e
    V_J = V * const.e

    # calcular vectores de onda (k1 y k2)
    hbar = const.hbar  # Constante de Planck reducida
    k1 = math.sqrt(2 * m * E_J) / hbar
    k2 = math.sqrt(2 * m * (E_J - V_J)) / hbar

    # Calcular probabilidades
    T = (4 * k1 * k2) / (k1 + k2)**2
    R = ((k1 - k2) / (k1 + k2))**2

    return T, R

# Parámetros del problema
m_electron = 9.11e-31  # kg
energia = 10.0  # eV
potencial = 9.0  # eV

# Calcular probabilidades
T, R = calcular_probabilidades_cuanticas(energia, potencial, m_electron)

# Imprimir resultados
print("Resultados para una partícula cuántica:")
print(f"- Energía de la partícula: {energia} eV")
print(f"- Altura del escalón de potencial: {potencial} eV")
print(f"- Masa de la partícula: {m_electron} kg (masa del electrón)")
print("\nProbabilidades calculadas:")
print(f"Transmisión (T): {T:.6f} ({T*100:.2f}%)")
print(f"Reflexión (R): {R:.6f} ({R*100:.2f}%)")
print(f"Suma T + R: {T+R:.6f} (debe ser 1 para conservación de probabilidad)")
```

5) calculo con las leyes de kepler

```python
import math
import scipy.constants as const

def calcular_parametros_orbitales(l1, v1):
    """
    calcula los parametros orbitales a partir de la distancia y velocidad en el perihelio

    Args:
        l1 (float): distancia en el perihelio (m)
        v1 (float): velocidad en el perihelio (m/s)

    Returns:
        dict: diccionario con los parámetros orbitales calculados
    """
    # constantes físicas
    G = 6.6738e-11  # Constante gravitacional (m³ kg⁻¹ s⁻²)
    M = 1.9891e30   # Masa del Sol (kg)

    # calcular v2 (la raíz más pequeña de la ecuación cuadratica)
    A = 1
    B = -2 * G * M / (v1 * l1)
    C = -(v1**2 - 2 * G * M / l1)

    discriminante = B**2 - 4 * A * C
    v2_1 = (-B + math.sqrt(discriminante)) / (2 * A)
    v2_2 = (-B - math.sqrt(discriminante)) / (2 * A)
    v2 = min(v2_1, v2_2)  # Tomamos la raíz más pequeña

    # Calcular l2 usando la segunda ley de kepler
    l2 = l1 * v1 / v2

    # Calcular otros parámetros orbitales
    a = 0.5 * (l1 + l2)  # Semieje mayor
    b = math.sqrt(l1 * l2)  # Semieje menor
    T = (2 * math.pi * a * b) / (l1 * v1)  # Periodo orbital en segundos
    T_anios = T / (60 * 60 * 24 * 365.25)  # Convertir a años
    e = (l2 - l1) / (l2 + l1)  # Excentricidad

    return {
        'afelio_distancia': l2,
        'afelio_velocidad': v2,
        'semieje_mayor': a,
        'semieje_menor': b,
        'periodo_orbital': T_anios,
        'excentricidad': e
    }

def main():
    print("Cálculo de parámetros orbitales según las leyes de Kepler")
    print("------------------------------------------------------")

    # Datos de prueba (Tierra y cometa Halley)
    datos_cuerpos = {
        'Tierra': {
            'l1': 1.4710e11,
            'v1': 3.0287e4
        },
        'Halley': {
            'l1': 8.7830e10,
            'v1': 5.4529e4
        }
    }

    # Calcular para cada cuerpo
    for nombre, datos in datos_cuerpos.items():
        print(f"\nResultados para {nombre}:")
        print(f"- Distancia en perihelio (ℓ1): {datos['l1']:.4e} m")
        print(f"- Velocidad en perihelio (v1): {datos['v1']:.4e} m/s")

        resultados = calcular_parametros_orbitales(datos['l1'], datos['v1'])

        print("\nParámetros orbitales calculados:")
        print(f"Distancia en afelio (ℓ2): {resultados['afelio_distancia']:.4e} m")
        print(f"Velocidad en afelio (v2): {resultados['afelio_velocidad']:.4e} m/s")
        print(f"Semieje mayor (a): {resultados['semieje_mayor']:.4e} m")
        print(f"Semieje menor (b): {resultados['semieje_menor']:.4e} m")
        print(f"Periodo orbital (T): {resultados['periodo_orbital']:.2f} años")
        print(f"Excentricidad orbital (e): {resultados['excentricidad']:.6f}")

if __name__ == "__main__":
    main()
```

7) polinomios de legendre

```python
import sympy as sp
x=sp.symbols('x')
def rodrigues(n):
  """calcula el polinomio de legendre con la formula de rodriguez"""
  return (1/(2**n * sp.factorial(n))) * sp.diff((x**2 - 1)**n, x, n)
p0=rodrigues(0)
p1=rodrigues(1)
print(f'Po(x)={sp.simplify(p0)}')
print(f'P1(x)={sp.simplify(p1)}')
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sympy import symbols, expand

x = symbols('x')

def legendre_recurrence(n):
    """
    Calcula el polinomio de Legendre Pₙ(x) usando relación de recurrencia
    """
    if n == 0:
        return 1
    elif n == 1:
        return x
    else:
        # Inicializamos los primeros dos polinomios
        P_prev_prev = 1  # P₀(x)
        P_prev = x       # P₁(x)

        for k in range(1, n):

            P_current = ((2*k + 1)*x*P_prev - k*P_prev_prev)/(k + 1)

            P_prev_prev, P_prev = P_prev, P_current

        return expand(P_prev)


polinomios = [legendre_recurrence(n) for n in range(1, 6)]


x_vals = np.linspace(-1, 1, 400)
P_funcs = [sp.lambdify(x, P, 'numpy') for P in polinomios]


plt.figure(figsize=(10, 6))
colors = ['b', 'g', 'r', 'c', 'm']

for i, (P_func, color) in enumerate(zip(P_funcs, colors), start=1):
    y_vals = P_func(x_vals)
    plt.plot(x_vals, y_vals, color, label=f'P_{i}(x)')


plt.title('Polinomios de Legendre para n=1 a 5', fontsize=14)
plt.xlabel('x', fontsize=12)
plt.ylabel('Pₙ(x)', fontsize=12)
plt.grid(True, linestyle='--', alpha=0.6)
plt.axhline(0, color='black', linewidth=0.5)
plt.axvline(0, color='black', linewidth=0.5)
plt.legend(loc='best')
plt.xlim(-1, 1)
plt.ylim(-1, 1.1)


for i, P in enumerate(polinomios, start=1):
    print(f"P_{i}(x) = {P}")

plt.tight_layout()
plt.show()
```
