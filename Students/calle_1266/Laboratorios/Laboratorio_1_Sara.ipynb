````python
#Función para saber si un número es defectivo
def números_defectivos(n):
  suma = 0
  for i in range(1, n): #hasta n - 1 ya que no consideramos a n para los divisores propios
    if n % i == 0: #si n es divisor propio se actualiza la variable suma
      suma += i
  if suma < n: #verificamos si es un número defectivo
    return True
  else:
    return False
````

````python
#Función para saber si un número es abundante
def números_abundantes(n):
  suma = 0
  for i in range(1, n): #hasta n - 1 ya que no consideramos a n para los divisores propios
    if n % i == 0: #si n es divisor propio se actualiza la variable suma
      suma += i
  if suma > n: #verificamos si es un número abundante
    return True
  else:
    return False
````

````python
#Función para saber si un número es semiperfecto
def números_semiperfectos(n):
  divisores = []
  for i in range(1, n):
    if n % i == 0: #si n es divisor propio se actualiza la variable suma
      divisores.append(i)

    def suma_divisores(k, suma): #función para encontrar las posibles combinaciones de las sumas de los divisores
      if suma == n:  # Si hay una suma que da exactamente a es semiperfecto
        return True
      if suma > n or k == len(divisores): #si la suma es mayor que el número o si ya se ha probado todos los divisores no es semiperfecto
        return False
      return suma_divisores(k + 1, suma) or suma_divisores(k + 1, suma + divisores[k]) #para incluir el divisor actual en la suma o no

  return suma_divisores(0, 0)
````

````python
números_semiperfectos(28)
````

````python
#función para saber si un número es perfecto
def números_perfectos(n):
  suma = 0
  for i in range(1, n): #hasta n - 1 ya que no consideramos a n para los divisores propios
    if n % i == 0: #si n es divisor propio se actualiza la variable suma
      suma += i
  return suma == n #verificamos si es un número perfecto
````

````python
números_perfectos(28)
````

````python
#función para saber si un número es primo
def números_primos(n):
  if n < 2: #2 es el primer número primo
    return False
  for i in range (2, n):
    if n % i == 0:
      return False
  return True
````

````python
#función para los números de catalan
def catalan(n):
  if n == 0:
    return 1
  if n > 0:
    return (4 * n - 2) / (n + 1) * catalan(n - 1)
  else:
    return "Ingrese un número positivo"

print(f'C_50 ={catalan(50)}')
````

````python
#Funcion para el máximo común divisor
def g(m, n):
  if n == 0:
    return m
  if n > 0:
    return g(n, m % n)
  else:
   return "Ingrese un número positivo"

print(f'El máximo común divisor de 108 y 192 es {g(108, 192)}')
````

Un problema conocido de mecánica cuántica involucra una partícula de masa $m$ que encuentra un escalón de potencial unidimensional, como se muestra en la figura:

image.png


La partícula con energía cinética inicial $E$ y vector de onda $k_1 = \sqrt{2mE}/\hbar$ entra desde la izquierda y encuentra un salto abrupto en la energía potencial de altura $V$ en la posición $x = 0$. Resolviendo la ecuación de Schrödinger, se puede demostrar que cuando $E > V$ la partícula puede:
- Pasar el escalón, en cuyo caso tendrá una energía cinética menor $E - V$ al otro lado y un vector de onda correspondientemente más pequeño $k_2 = \sqrt{2m(E - V)}/\hbar$
- Ser reflejada, manteniendo toda su energía cinética y vector de onda sin cambios, pero moviéndose en dirección opuesta.


Las probabilidades $T$ y $R$ para transmisión y reflexión están dadas por:

$$T = \frac{4k_1k_2}{(k_1 + k_2)^2}, \quad R = \left(\frac{k_1 - k_2}{k_1 + k_2}\right)^2.$$

Consideremos una partícula con masa igual a la del electrón $m = 9.11 \times 10^{-31} \, \text{kg}$ y energía $10 \, \text{eV}$ que encuentra un escalón de potencial de altura $9 \, \text{eV}$. Escriba un programa en Python para calcular e imprimir las probabilidades de transmisión y reflexión usando las fórmulas anteriores.


````python
import numpy as np
````

````python
#definimos las variables
m = 9.11e-31
h = 6.626e-34
h_barra= h/2 * np.pi
Ev_joule = 1.602e-19
E = 10 * Ev_joule
V = 9 * Ev_joule
````

````python
#definimos las funciones para la probabilidad
def prob_T(m, E, V, h_barra):
  k_1 = np.sqrt((2*m*E))/(h_barra)
  k_2 = np.sqrt(2*m*(E-V))/(h_barra)
  T = (4*k_1*k_2)/((k_1+k_2)**2)
  return T
````

````python
def prob_R(m, E, V,h_barra):
  k_1 = np.sqrt((2*m*E))/(h_barra)
  k_2 = np.sqrt(2*m*(E-V))/(h_barra)
  R= ((k_1-k_2)/(k_1+k_2))**2
  return R
````

````python
suma = prob_T(m,E,V,h_barra) + prob_R(m,E,V,h_barra)
````

````python
print(f'La probabilidad de Transmisión es:{prob_T(m,E,V,h_barra)}')
print(f'La probabilidad de Reflexión es:{prob_R(m,E,V,h_barra)}')
print(f'La suma de las probabilidades es:{suma}')
````

La órbita en el espacio de un cuerpo alrededor de otro, como un planeta alrededor del Sol, no necesita ser circular. En general toma la forma de una elipse, con el cuerpo a veces más cerca y a veces más lejos. Si se conocen la distancia $\ell_{1}$ del acercamiento más próximo que hace un planeta al Sol, llamado perihelio, y su velocidad lineal $v_{1}$ en el perihelio, entonces cualquier otra propiedad de la órbita puede calcularse a partir de estos dos datos de la siguiente manera.
La segunda ley de Kepler nos dice que la distancia $\ell_{2}$ y velocidad $v_{2}$ del planeta en su punto más distante, o afelio, satisfacen $\ell_{2}v_{2}=\ell_{1}v_{1}$. Al mismo tiempo, la energía total, cinética más gravitacional, de un planeta con velocidad $v$ y distancia $r$ del Sol está dada por
    $$
    E = \tfrac{1}{2}mv^{2} - G\frac{mM}{r},
    $$
    donde $m$ es la masa del planeta, $M=1.9891\times 10^{30}\,\mathrm{kg}$ es la masa del Sol, y $G=6.6738\times 10^{-11}\,\mathrm{m}^{3}\,\mathrm{kg}^{-1}\,\mathrm{s}^{-2}$ es la constante gravitacional de Newton. Dado que la energía debe conservarse, se llega a que $v_{2}$ es la raíz más pequeña de la ecuación cuadrática
    $$
    v_{2}^{2} - \frac{2GM}{v_{1}\ell_{1}}v_{2} - \left[v_{1}^{2} - \frac{2GM}{\ell_{1}}\right] = 0.
    $$
    Una vez que se tiene $v_{2}$, podemos calcular $\ell_{2}$ usando la relación $\ell_{2}=\ell_{1}v_{1}/v_{2}$.

- Dados los valores de $v_{1}$, $\ell_{1}$, y $\ell_{2}$, otros parámetros de la órbita vienen dados por fórmulas simples que pueden derivarse de las leyes de Kepler y del hecho de que la órbita es una elipse:
$$
    \text{Semieje mayor:}  \quad a = \tfrac{1}{2}(\ell_{1}+\ell_{2}),
$$
$$
    \text{Semieje menor:}  \quad b = \sqrt{\ell_{1}\ell_{2}},
$$
$$
    \text{Periodo orbital:}  \quad T = \frac{2\pi ab}{\ell_{1}v_{1}},
$$
$$
    \text{Excentricidad orbital:}  \quad e = \frac{\ell_{2}-\ell_{1}}{\ell_{2}+\ell_{1}}.
$$

Escriba un programa que solicite al usuario ingresar la distancia al Sol y la velocidad en el perihelio, luego calcule e imprima las cantidades $\ell_{2}$, $v_{2}$, $T$, y $e$.

- Pruebe su programa calculando las propiedades de las órbitas de la Tierra (para la cual $\ell_{1}=1.4710\times 10^{11}\,\mathrm{m}$ y $v_{1}=3.0287\times 10^{4}\,\mathrm{m}\,\mathrm{s}^{-1}$) y del cometa Halley ($\ell_{1}=8.7830\times 10^{10}\,\mathrm{m}$ y $v_{1}=5.4529\times 10^{4}\,\mathrm{m}\,\mathrm{s}^{-1}$). Entre otras cosas, debería encontrar que el período orbital de la Tierra es un año y el del cometa Halley es alrededor de 76 años.


````python
import math
#definimos las variables
G = 6.6738e-11   #constante gravitacional en m^3 kg^-1 s^-2
M = 1.9891e30    #masa del Sol en kg

#definimos la función para calcular los parámetros orbitales
def calcular_orbita(l1, v1):
    """
    Calcula y muestra v2, l2, semiejes, período orbital y excentricidad
    a partir de la distancia al Sol y velocidad en el perihelio.
    """

    #coeficientes de la ecuación cuadrática para v2
    A = 1
    B = -(2 * G * M) / (v1 * l1)
    C = -v1**2 + (2 * G * M) / l1

    #discriminante de la ecuación cuadrática
    discriminante = B**2 - 4 * A * C

    #verificamos que haya solución real
    if discriminante < 0:
        print("No hay solución real para v2.")
        return

    # Calculamos las dos soluciones de la cuadrática
    D = math.sqrt(discriminante)
    v2_1 = (-B + D) / (2 * A)
    v2_2 = (-B - D) / (2 * A)

    #elegimos la raíz positiva menor, ya que esta es la solución
    v2 = min(v2_1, v2_2)

    #calculamos l2
    l2 = l1 * v1 / v2

    #calculamos los demás parámetros orbitales
    a = 0.5 * (l1 + l2)               # Semieje mayor
    b = math.sqrt(l1 * l2)           # Semieje menor
    T = (2 * math.pi * a * b) / (l1 * v1)  # Período orbital
    e = (l2 - l1) / (l2 + l1)         # Excentricidad orbital

    #imprimimos los resultados
    print("\nResultados calculados:")
    print(f"Velocidad en el afelio (v2): {v2:.3e} m/s")
    print(f"Distancia en el afelio (l2): {l2:.3e} m")
    print(f"Semieje mayor (a): {a:.3e} m")
    print(f"Semieje menor (b): {b:.3e} m")
    #print(f"Período orbital (T): {T:.3e} segundos")
    print(f"Período orbital (T): {T / (60*60*24*365.25):.2f} años")
    print(f"Excentricidad orbital (e): {e:.3f}")

#solicitamos los datos al usuario
print("Cálculo de parámetros orbitales")
l1 = float(input("Ingresa la distancia al Sol en el perihelio (l1) en metros: "))
v1 = float(input("Ingresa la velocidad en el perihelio (v1) en metros por segundo: "))

#llamamos la función con los valores dados
calcular_orbita(l1, v1)

#para la tierra
calcular_orbita(1.4710e11, 3.0287e4)

#para el cometa Halley
calcular_orbita(8.7830e10, 5.4529e4)
````

Graficar la trayectoria del proyectil en su plano de movimiento (considerando el eje $x$ paralelo a la horizontal y el origen en la posición inicial del proyectil) desde el punto de lanzamiento hasta el punto donde impacta el suelo (asumiendo un terreno plano). La gráfica debe incluir etiquetas en los ejes y un título. Recuerde que las ecuaciones del movimiento parabólico están dadas por

$$
\text{Posición horizontal:} \quad x(t) = v_{0x} t = v_0 \cos\theta \cdot t
$$
$$
\text{Posición vertical:} \quad y(t) = y_0 + v_{0y} t - \frac{1}{2}gt^2 = y_0 + v_0 \sin\theta \cdot t - \frac{1}{2}gt^2
$$
$$
\text{Velocidad horizontal:} \quad v_x(t) = v_{0x} =v_0 \cos \theta= \text{constante}
$$
$$
\text{Velocidad vertical:} \quad v_y(t) = v_{0y} - gt = v_0 \sin\theta - gt
$$

````python
import matplotlib.pyplot as plt

#definimos las variables
y0= 0 #en m
v0= 4.0 #en m
theta= 45*(np.pi/180) #pasamos de grados a radianes
g= 9.8 #en m/s^2
v0x= v0 * np.cos(theta) #velocidad inicial en la componente x
v0y= v0 * np.sin(theta) #velocidad inicial en la componente y

t_vuelo = (2 * v0y)/g #despejamos t de la posición vertical
t= np.linspace(0, t_vuelo, 100) #para graficar

x = v0x * t #posición horizontal
y= y0 + v0y * t - ((1/2)*g*t**2) #posición vertical

plt.figure(figsize=(10,6))
plt.plot(x, y, color = 'purple', label='Trayectoria del proyectil')
plt.title('Trayectoria de un proyectil')
plt.xlabel('Posición horizontal')
plt.ylabel('Posición vertical')
plt.grid(True)
plt.axhline(0, color ='pink', linewidth=0.5)
plt.legend()
plt.show()
````

# 7.

Los polinomios de Legendre son de gran importancia en física, especialmente en la solución de la ecuación de Laplace en coordenadas esféricas. Son fundamentales en la expansión en armónicos esféricos, lo que los hace esenciales en problemas de mecánica cuántica y electromagnetismo. Cada polinomio de Legendre $P_n(x)$ es un polinomio de grado $n$. Este puede ser expresado usando la Fórmula de Rodrigues:
     
$$
P_n (x)=\frac{1}{n!\ 2^n} \frac{d^n}{dx^n} (x^2-1)^n, \ \ n \in  \mathbb{Z^{+}}+\{0\}
$$
        
- Utilizando la fórmula de Rodrigues halle $P_0(x)$ y $P_1(x)$.
-  Con la fórmula de recurrencia
$$
            P_{j+1}(x)=\frac{(2j+1)x P_j (x)-jP_{j-1}(x)}{j+1},
$$
elabore un código que permita calcular $P_n(x)$ para $n\geq2$.
- En una misma gráfica muestre $P_n(x)$ para $1\leq n\leq5$.


````python
def legendre(n, x):
  #caso base: P_0(x) = 1
  if n == 0:
    return 1
  #caso base: P_1(x) = x
  elif n == 1:
    return x

  #inicializamos P_0(x) y P_1(x)
  pj = x
  pj_ = 1
  for i in range(1, n): #iteramos desde j = 1 hasta n - 1
    #aplicamos la fórmula de recurrencia
    pjn = ((2*i+1)*x*pj - i*pj_)/(i+1)
    #para actualizar la variable
    pj_ = pj
    pj = pjn
  return pjn
````

````python
x = np.linspace(-1, 1, 100)
n = [1, 2, 3, 4, 5] #lista de los grados de los polinomios que queremos graficar

for n_ in n: #recorremos cada grado de polinomio
  P_value = [] #lista vacía para guardar los valores de P_n(x)
  for i in x: #para cada valor de x en el intervalo [-1, 1]
    P = legendre(n_, i) #calculamos el valor del polinomio de Legendr
    P_value.append(P) #guardamos el resultado en la lista
  plt.plot(x, P_value, label=f'{n_}') #graficamos P_n(x)
  plt.legend()
  plt.grid(True)
````
