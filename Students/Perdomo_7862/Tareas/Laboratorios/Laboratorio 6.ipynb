# 3.
Se sabe que un sistema físico es gobernado por las ecuaciones,

$$
\begin{matrix}
x' &=& 2x&-y&+z, \\
y' &=&   &3y&-z, \\
z' &=& 2x&+y&+3z,
\end{matrix}
$$

donde $x',y',z'$ son las derivadas, en forma vectorial esto se puede representar como $\mathbf{r}'=A\mathbf{r}$ que tiene solución analítica general, $\mathbf{r} = e^{At} \mathbf{r}(0)$.

- Realice la diagonalización de $A$.

- Use el hecho de que $f(A) =U f(D) U^{-1}$ para encontrar $x(t),y(t),z(t)$ en el intervalo de tiempo $[0,2]$.

- Grafique $x(t),y(t),z(t)$ (use la condición inicial $\mathbf{r}(0)=(1,1,2))$.

- Compare el resultado a la solución analítica,

$$
2x  = x(0)(e^{2t}(1+e^{2t}-2t)) + y(0) (-2te^{2t}) + z(0)(e^{2t}(-1+e^{2t})), 
$$

$$
2y  = x(0)(-e^{2t}(-1+e^{2t}-2t)) + y(0)(2(t+1)e^{2t}) + z(0)(-e^{2t}(-1+e^{2t})),
$$

$$
2z  = x(0)(e^{2t}(-1+e^{2t}+2t)) + y(0)(2te^{2t}) + z(0)(e^{2t}(1+e^{2t})).
$$

```python
#se escribe la matríz A 
A=np.array([
    [2,-1,1],
    [0,3,-1],
    [2,1,3]
])
#se calculan los valores y vectores propios
valores,vectores=la.eig(A)
#se realiza la diagonalización
diagonal=np.diag(valores)
#se define r0
r_0=np.array([1,1,2])
#se calcula la inversa de vectores
vec_inv=la.inv(vectores)
#se realiza la función r(t)
def r(t,d,u,u_inv,r0):
    """
    Calcula el vector desplazamiento de un sistema dado
    Entradas:
        t: tiempo -> float
        d: matriz diagonal -> ndarray
        u: Matriz de vectores propios de A -> ndarray
        u_inv: Matriz inversa de u -> ndarray
        r0: condición inicial -> array
    Salidas:
        r_t: vector posición en el instante t -> array
    """ 
    #se calcula la evaluación de f en D (elevando cada autovalor)
    f_d = np.diag(np.exp(np.diag(d) * t))
    #se calcula el producto matricial con la condición inicial
    r_t = u @ f_d @ u_inv @ r0
    return r_t
#se define la solución exacta
def sol_exacta(t,v):
    """
    Implementa la solución exacta al sistema físico dado
    Entradas:
        t: float -> tiempo del sistema físico 
        v: vector inicial (x0,y0,z0) -> array
    Salidas:
        x: Componente x del vector posición en t -> float
        y: Componente y del vector posición en t -> float
        z: Componente z del vector posición en t -> float
    """
    #se saca cada componente del vector inicial (x0,y0,z0)
    x0,y0,z0=v
    #se calcula e**2t
    exp=np.exp(2*t)
    #se calcula cada componente
    x=0.5*(x0*(exp*(1+exp-2*t))+y0*(-2*t*exp)+z0*(exp*(-1+exp)))
    y=0.5*(x0*(-exp*(-1+exp-2*t))+y0*(2*(t+1)*exp)+z0*(-exp*(-1+exp)))
    z=0.5*(x0*(exp*(-1+exp+2*t))+y0*(2*t*exp)+z0*(exp*(1+exp)))
    #se crea el vector r
    r=np.array([x,y,z])
    return r
#se define el intervalo de tiempo
t_values=np.linspace(0,2,20)
#se inicializan varias listas
r_exacts=[]
r_nums=[]
errores=[]
for t in t_values:
    #se determina la solución exacta
    r_exact=sol_exacta(t,r_0)
    #se determina la solución numérica
    r_num=r(t,diagonal,vectores,vec_inv,r_0)
    #se calcula el error
    error=la.norm(r_exact-r_num)
    errores.append(error)
    r_exacts.append(r_exact)
    r_nums.append(r_num)
#se convierten a matrices
r_exacto=np.array(r_exacts)
r_numerico=np.array(r_nums)
#se hace la gráfica
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

# Vectores exactos (color rojo) con etiqueta solo en el primero
ax.quiver(0, 0, 0,
          r_exacto[0,0], r_exacto[0,1], r_exacto[0,2],
          color='red', arrow_length_ratio=0.1, label='Solución exacta')
for i in range(1, r_exacto.shape[0]):
    ax.quiver(0, 0, 0,
              r_exacto[i,0], r_exacto[i,1], r_exacto[i,2],
              color='red', arrow_length_ratio=0.1)

# Vectores numéricos (color azul) con etiqueta solo en el primero
ax.quiver(0, 0, 0,
          r_numerico[0,0], r_numerico[0,1], r_numerico[0,2],
          color='blue', arrow_length_ratio=0.1, label='Solución numérica')
for i in range(1, r_numerico.shape[0]):
    ax.quiver(0, 0, 0,
              r_numerico[i,0], r_numerico[i,1], r_numerico[i,2],
              color='blue', arrow_length_ratio=0.1)

#se declara una escala
escala=2400
ax.set_xlim([-escala,escala ])
ax.set_ylim([-escala, escala])
ax.set_zlim([-escala, escala])
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')

#se coloca la leyenda
ax.legend()

plt.show()
#se grafica el error
plt.figure(figsize=(8,5))
plt.plot(t_values, errores)
plt.title("Errores", fontsize=20)
plt.xlabel("tiempo",fontsize=15)
plt.ylabel("Error", fontsize=15)
plt.grid()
plt.show()
```

Un electrón interactúa con un campo magnético uniforme. Para determinar los niveles de energía permitidos del electrón en presencia del campo magnético, es necesario resolver la ecuación:

$$
\hat{H}|\Psi\rangle = E|\Psi\rangle
$$

donde el Hamiltoniano está dado por $H = -\bm{\mu} \cdot \bm{B} = -\gamma \bm{B} \cdot \bm{S}$, siendo $\gamma$ la razón giromagnética, $\bm{B}$ el campo magnético y $\bm{S}$ el espín. Puede demostrarse que el Hamiltoniano se expresa como:

$$
\hat{H} = -\frac{\gamma\hbar}{2}\begin{pmatrix} 
B_z & B_x - iB_y \\ 
B_x + iB_y & -B_z 
\end{pmatrix} 
$$

El valor experimental de $\gamma$ para el electrón es 2. El orden de magnitud del campo magnético es 1 gauss en dirección z.

- Verifique que $H$ es hermítica.
- Encuentre los niveles de energía permitidos.
- Determine los autovectores y verifique su normalización.



```python
import numpy as np
from numpy.linalg import eigh, norm
from scipy.constants import hbar, e  # hbar en J·s, e en C (solo por si quieres pasar a eV)

# Parámetros (puedes cambiarlos)
gamma = 2.0           # valor "experimental" dado en el enunciado (adimensional aquí según la forma dada)
B = np.array([0.0, 0.0, 1.0])   # campo en gauss (orden de magnitud 1 G en dirección z)
G_TO_T = 1e-4         # 1 gauss = 1e-4 tesla
B_SI = B * G_TO_T     # convertir a tesla

# Construcción de la matriz Hamiltoniana H = - γ ħ/2 * [ [ Bz, Bx - i By ], [ Bx + i By, -Bz ] ]
Bx, By, Bz = B_SI
prefactor = - gamma * hbar / 2.0
H = prefactor * np.array([[Bz, Bx - 1j*By],
                          [Bx + 1j*By, -Bz]], dtype=complex)

# 1) Verificar hermiticidad: H^\dagger == H (numéricamente dentro de una tolerancia)
is_hermitian = np.allclose(H.conj().T, H, atol=1e-12)
print("¿H es hermítica? ->", is_hermitian)
if not is_hermitian:
    print("Máxima diferencia entre H^† y H:", np.max(np.abs(H.conj().T - H)))

# 2) Autovalores (energías) y autovectores
vals, vecs = eigh(H)   # eigh retorna autovalores ordenados y autovectores (columnas)
print("\nAutovalores (energías) en J:")
print(vals)

# también los mostramos en eV para mayor intuición
J_to_eV = 1.0 / e
print("\nAutovalores (energías) en eV (aprox):")
print(vals * J_to_eV)

# 3) Analítica para B=(0,0,Bz): E_± = ∓ γ ħ Bz / 2
E_plus_analytic  = - gamma * hbar * Bz / 2.0   # signo según la expresión (fíjate en el prefactor)
E_minus_analytic = + gamma * hbar * Bz / 2.0
print("\nValores analíticos previstos (J):")
print("E_+  =", E_plus_analytic)
print("E_-  =", E_minus_analytic)

# 4) Autovectores y normalización
print("\nAutovectores (columnas) retornados por eigh :")
print(vecs)

# Comprobación de normalización (norma 2 = 1)
norms = np.array([norm(vecs[:, i]) for i in range(vecs.shape[1])])
print("\nNormas de cada autovector :", norms)

# Comprobación de ortogonalidad (vecs.conj().T @ vecs ≈ I)
orto = vecs.conj().T @ vecs
print("\nProducto interno de autovectores:\n", orto)

# 5) Relacionar autovalores con autovectores verificando H v = E v (consistencia numérica)
residuos = []
for i in range(len(vals)):
    lhs = H @ vecs[:, i]
    rhs = vals[i] * vecs[:, i]
    residuos.append(norm(lhs - rhs))
print("\nResiduos ||H v - E v|| para cada modo:", residuos)
```

Considere el sistema de tres masas y cuatro resortes mostrado en la figura.

Las ecuaciones diferenciales que describen el movimiento del sistema masa-resorte son:

$$
k \left( -2u_1 + u_2 \right) = m\ddot{u}_1 
$$

$$
k(u_1 - 2u_2 + u_3) = 3m\ddot{u}_2 
$$

$$
k(u_2 - 2u_3) = 2m\ddot{u}_3
$$

donde:
- $u_i(t)$ representa el desplazamiento de la masa $i$ desde su posición de equilibrio.
- $k$ es la constante de los resortes.
- $m$ es la masa.

Determinar las frecuencias de vibración ($\omega$) en términos de $\sqrt{k/m}$ y las formas modales correspondientes.

```python
def calcular_frecuencias_masa_resorte(k, masas):
    """
    Calcula las frecuencias más bajas y los modos de vibración para un sistema masa-resorte.
    
    Parámetros:
    k - constante del resorte (en N/m)
    masas - array de masas (en kg)
    
    Retorna:
    frecuencias - array de frecuencias (en rad/s) ordenadas de menor a mayor
    modos - matriz donde cada columna es un modo de vibración normalizado
    """
    n = len(masas)
    
    # Construir la matriz K
    K = np.zeros((n, n))
    for i in range(n):
        if i == 0:
            K[i, i] = 2*k
            K[i, i+1] = -k
        elif i == n-1:
            K[i, i] = k
            K[i, i-1] = -k
        else:
            K[i, i-1] = -k
            K[i, i] = 2*k
            K[i, i+1] = -k
    
    # Construir la matriz de masa M (diagonal)
    M = np.diag(masas)
    
    # Resolver el problema generalizado de valores propios: K v = w² M v
    valores_propios, modos = la.eigh(K, M)
    
    # Calcular frecuencias en rad/s 
    frecuencias = np.sqrt(valores_propios) 
    
    return frecuencias, modos
```
