```python
def numero_defectivo(n):
    """
    Determina si un número es defectivo.

    Un número se considera defectivo si la suma de sus divisores propios 
    (excluyendo el número mismo) es menor que el propio número.

    Parámetro:
    ----------
    n : int
        Número entero positivo a evaluar.

    Retorno:
    --------
    bool
        True si el número es defectivo, False en caso contrario.
    """
    suma_divisores = 0
    for i in range(1, n):
        if n % i == 0:
            suma_divisores += i
    if suma_divisores < n:
        return True
    else:
        return False
```

```python
def numeros_abundantes(n):
    """
    Determina si un número es abundante.

    Un número es abundante si la suma de sus divisores propios (excluyendo el número mismo)
    es mayor que el propio número.

    Parámetro:
    ----------
    n : int
        Número entero positivo a evaluar.

    Retorno:
    --------
    bool
        True si el número es abundante, False en caso contrario.

    Ejemplo:
    --------
    >>> numeros_abundantes(12)
    True
    >>> numeros_abundantes(9)
    False
    """
    suma_divisores=0
    for i in range(1,n):
        if n%i==0:
            suma_divisores+=i
        if suma_divisores>n:
            return True
        else: 
            return False
```

```python
def numeros_semiperfectos(n):
    """
    Determina si un número es semiperfecto usando Python puro.

    Un número es semiperfecto si existe una combinación de sus divisores propios
    cuya suma es exactamente el número.

    Parámetro:
    ----------
    n : int
        Número entero positivo a evaluar.

    Retorno:
    --------
    bool
        True si el número es semiperfecto, False en caso contrario.
    """
    divisores = [i for i in range(1, n) if n % i == 0]
    
    def encontrar_suma(subset, suma_actual, indice):
        """Función recursiva para buscar subconjuntos con suma igual a n."""
        if suma_actual == n:
            return True
        if suma_actual > n or indice >= len(divisores):
            return False

        # Elegir el elemento actual o ignorarlo
        return (encontrar_suma(subset + [divisores[indice]], suma_actual + divisores[indice], indice + 1) or
                encontrar_suma(subset, suma_actual, indice + 1))

    return encontrar_suma([], 0, 0)
```

```python
def numeros_perfectos(n):
    """
    Determina si un número es perfecto.

    Un número es perfecto si la suma de sus divisores propios (excluyendo el mismo número)
    es exactamente igual al número.

    Parámetro:
    ----------
    n : int
        Número entero positivo a evaluar.

    Retorno:
    --------
    bool
        True si el número es perfecto, False en caso contrario.

    Ejemplo:
    --------
    >>> numeros_perfectos(6)
    True
    >>> numeros_perfectos(10)
    False
    """
    suma_divisores = 0
    for i in range(1, n):
        if n % i == 0:
            suma_divisores += i
    if suma_divisores == n:
        return True
    else:
        return False
```

```python
def numeros_primos(n):
    """
    Determina si un número es primo.

    Un número es primo si solo tiene dos divisores: 1 y él mismo.

    Parámetro:
    ----------
    n : int
        Número entero positivo a evaluar.

    Retorno:
    --------
    bool
        True si el número es primo, False en caso contrario.

    Ejemplo:
    --------
    >>> numeros_primos(7)
    True
    >>> numeros_primos(10)
    False
    """
    contador_divisores = 0
    for i in range(1, n + 1):
        if n % i == 0:
            contador_divisores += 1
        if contador_divisores > 2:  # Si tiene más de dos divisores, no es primo
            return False
    return contador_divisores == 2
```

## **Punto 2**

Determine si un número n entero ingresado por el usuario es un palíndromo, (Retorne True en caso afirmativo y False en caso contrario)

```python
def numero_palindromo(n):
    """
    Determina si un número es palíndromo.

    Un número es palíndromo si se lee igual de izquierda a derecha que de derecha a izquierda.

    Parámetro:
    ----------
    n : int
        Número entero positivo a evaluar.

    Retorno:
    --------
    bool
        True si el número es palíndromo, False en caso contrario.

    """
    if str(n) == str(n)[::-1]:
        return True
    else:
        return False
```

```python
def catalan(n):
    """
    Calcula el número de Catalan C_n usando recursión.

    Parámetro:
    ----------
    n : int
        Índice del número de Catalan a calcular.

    Retorno:
    --------
    int
        Valor del número de Catalan C_n.
    """
    if n == 0:
        return 1
    return ((4 * n - 2) * catalan(n - 1)) // (n + 1)
```

```python
#Calcular Catalán 50
print(catalan(50))
```

```python
def mcd(m, n):
    """
    Calcula el Máximo Común Divisor (MCD) de dos enteros no negativos 
    utilizando el Algoritmo de Euclides de forma recursiva.

    Parámetros:
    -----------
    m : int
        Primer número entero no negativo.
    n : int
        Segundo número entero no negativo.

    """
    return m if n == 0 else mcd(n, m % n)
```

```python
import numpy as np

# Definimos constantes
m = 9.11e-31  # Masa del electrón en kg
hbar = 1.0545718e-34  # Constante reducida de Planck en J·s
E = 10 * 1.6e-19  # Energía en julios (10 eV)
V = 9 * 1.6e-19  # Potencial en julios (9 eV)

# Calculamos los números de onda
k1 = np.sqrt(2 * m * E) / hbar
k2 = np.sqrt(2 * m * (E - V)) / hbar

# Calculamos las probabilidades de transmisión y reflexión
T = (4 * k1 * k2) / ((k1 + k2) ** 2)
R = ((k1 - k2) / (k1 + k2)) ** 2

print(f"Probabilidad de transmisión: {T:.4f}")
print(f"Probabilidad de reflexión: {R:.4f}")
```

```python
import math

# Constantes
G = 6.6738e-11  # Constante gravitacional
M = 1.9891e30   # Masa del Sol

def calcular_orbita(l1, v1):
    # Resolver la cuadrática para v2
    A = 1
    B = -2 * G * M / (v1 * l1)
    C = - (v1**2 - (2 * G * M / l1))
    
    discriminante = B**2 - 4*A*C
    if discriminante < 0:
        raise ValueError("Discriminante negativo, no hay solución real para v2")

    v2 = (-B - math.sqrt(discriminante)) / (2*A)  # Raíz menor
    
    # Calcular l2
    l2 = l1 * v1 / v2

    # Semieje mayor
    a = 0.5 * (l1 + l2)

    # Semieje menor
    b = math.sqrt(l1 * l2)

    # Período orbital
    T = (2 * math.pi * a * b) / (l1 * v1)

    # Excentricidad
    e = (l2 - l1) / (l2 + l1)

    return l2, v2, a, b, T, e
```

```python
#Ejemplo planeta Tierra
l1 = 1.4710e11  # m
v1 = 3.0287e4   # m/s

l2, v2, a, b, T, e = calcular_orbita(l1, v1)

print(f"Distancia al afelio (l2): {l2:.2e} m")
print(f"Velocidad en el afelio (v2): {v2:.2e} m/s")
print(f"Semieje mayor (a): {a:.2e} m")
print(f"Semieje menor (b): {b:.2e} m")
print(f"Período orbital (T): {T/(60*60*24*365.25):.2f} años")
print(f"Excentricidad (e): {e:.3f}")
```

```python
#Ejemplo cometa Halley
l1 = 8.7830e10  # m
v1 = 5.4529e4   # m/s

l2, v2, a, b, T, e = calcular_orbita(l1, v1)

print(f"Distancia al afelio (l2): {l2:.2e} m")
print(f"Velocidad en el afelio (v2): {v2:.2e} m/s")
print(f"Semieje mayor (a): {a:.2e} m")
print(f"Semieje menor (b): {b:.2e} m")
print(f"Período orbital (T): {T/(60*60*24*365.25):.2f} años")
print(f"Excentricidad (e): {e:.3f}")
```

```python
import numpy as np
import matplotlib.pyplot as plt
y0=0   #altura inicial
v0=4.0 #velocidad inicial
theta_grados= 45 #ángulo
g=9.8
#convertir theta a radianes
theta_radianes=(theta_grados*np.pi)/180
#Determinar Seno y Coseno
cos_theta=np.cos(theta_radianes)
sen_theta=np.sin(theta_grados)
def x(t):
  """   Se ingresa un valor t (tiempo) y devuelve el desplazamiento en el eje x  """
  x=v0*t*cos_theta
  return x
def y(t):
  """   Se ingresa un valor t (tiempo) y devuelve el desplazamiento en el eje y  """
  y=y0+v0*t*sen_theta-(0.5*g*(t**2))
  return y
def formula_cuadratica (a,b,c):
  """   Se ingresan valores a,b,c correspondientes a los
  coeficientes del polinomio de segundo grado asociado a la
  ecuación del y(t) y devuelve solo las raíces positivas del polinomio,
  si no las hay, indica que las raíces son negativas o imaginarias   """
  t1=(-b+(b**2-4*a*c)**(0.5))/(2*a)
  t2=(-b-(b**2-4*a*c)**(0.5))/(2*a)
  if t1>0:
    return t1
  elif t2>0:
   return t2
  else:
    return "La solución es negativa o imaginaría"
#coeficientes del polinomio cuadrático asociado a y(t)
a=0.5*g
b=-v0*sen_theta
c=-y0
#Determinar el tiempo de vuelo con a,b,c
tiempo_vuelo =formula_cuadratica(a,b, c)
#Crear un conjunto de valores para realizar la gráfica
tiempo=np.linspace(y0,tiempo_vuelo, 100)
#Gráfica
fig=plt.figure(figsize=(12,5))
plt.plot(x(tiempo), y(tiempo))
plt.xlabel("x(t)", fontsize=20)
plt.ylabel("y(t)", fontsize=20)
plt.title("Trayectoria del Proyectil en X vs Y", fontsize=30)
plt.grid(alpha=0.6)
plt.axis("equal")
plt.axhline(0,color="red",linestyle="--")
plt.show ()
```

```python
def legendre(n, x):
    """Calcula el polinomio de Legendre P_n(x) usando la fórmula de recurrencia."""
    if n == 0:
        return np.ones_like(x)
    elif n == 1:
        return x
    else:
        Pn_1 = x
        Pn_2 = np.ones_like(x)
        for j in range(1, n):
            Pn = ((2*j + 1)*x*Pn_1 - j*Pn_2) / (j + 1)
            Pn_2, Pn_1 = Pn_1, Pn
        return Pn

# Crear los valores de x entre -1 y 1
x_vals = np.linspace(-1, 1, 400)

# Graficar P_n(x) para n = 1 a 5
plt.figure(figsize=(10, 6))
for n in range(1, 6):
    y_vals = legendre(n, x_vals)
    plt.plot(x_vals, y_vals, label=f'$P_{n}(x)$')

plt.title('Polinomios de Legendre $P_n(x)$ para $1 \\leq n \\leq 5$')
plt.xlabel('$x$')
plt.ylabel('$P_n(x)$')
plt.grid()
plt.legend()
plt.show()
```
