Diseñar un programa en el que entrado  un numero `a`  retorne una variable booleana True or false si cumple que es:

- [Números defectivo](https://es.wikipedia.org/wiki/Número_defectivo) : la suma de los divisores propios es menor que el número. 

  La rutina se deberá llamar números_defectivos



- [Números abundantes](https://es.wikipedia.org/wiki/Número_abundante): la suma de los divisores es mayor que el número.

  La rutina se deberá llamar números_abundantes



- [Números semiperfectos](https://es.wikipedia.org/wiki/N%C3%BAmero_semiperfecto) la suma de todos o algunos de los divisores propios es igual al número.

  La rutina se debera llamar numeros_semiperfectos

- [Números perfectos](https://es.wikipedia.org/wiki/N%C3%BAmero_perfecto) la suma de todos sus divisores propios, excepto el mismo numero, es igual al número.

   La rutina se deberá llamar numeros_perfectos


- [Números primos](https://es.wikipedia.org/wiki/N%C3%BAmero_primo) el número es divisible unicamente por sí mismo y por 1.

   La rutina se deberá llamar numeros_primos


**Problema no obligatorio** 

Retornar los 30 primeros números de cada clase

````python
# Funcion para obtener los divisores propios de un numero
def obtener_divisores_propios(numero):
    # Retorna una lista con todos los divisores propios de 'numero'
    # Los divisores propios son los divisores menores que el mismo numero
    divisores = []
    for i in range(1, numero):
        if numero % i == 0:
            divisores.append(i)
    return divisores

# Funcion que determina si un numero es defectivo
def numeros_defectivos(a):
    # Retorna True si la suma de los divisores propios es menor que el numero
    suma_divisores = sum(obtener_divisores_propios(a))
    return suma_divisores < a

# Funcion que determina si un numero es abundante
def numeros_abundantes(a):
    # Retorna True si la suma de los divisores propios es mayor que el numero
    suma_divisores = sum(obtener_divisores_propios(a))
    return suma_divisores > a

# Funcion que determina si un numero es perfecto
def numeros_perfectos(a):
    # Retorna True si la suma de los divisores propios es exactamente igual al numero
    suma_divisores = sum(obtener_divisores_propios(a))
    return suma_divisores == a

# Funcion que determina si un numero es primo
def numeros_primos(a):
    # Retorna True si el numero es primo
    # Un numero primo tiene solo dos divisores: 1 y el mismo
    if a <= 1:
        return False
    for i in range(2, int(a**0.5) + 1):
        if a % i == 0:
            return False
    return True

# Funcion que determina si un numero es semiperfecto
def numeros_semiperfectos(a):
    # Retorna True si una combinacion de sus divisores propios suma exactamente el numero
    divisores = obtener_divisores_propios(a)

    # Usamos backtracking para verificar si alguna combinacion suma a
    def buscar_combinacion(indice, suma_actual):
        if suma_actual == a:
            return True
        if suma_actual > a or indice >= len(divisores):
            return False
        # Intentar incluir el divisor actual
        if buscar_combinacion(indice + 1, suma_actual + divisores[indice]):
            return True
        # Intentar sin incluir el divisor actual
        return buscar_combinacion(indice + 1, suma_actual)

    return buscar_combinacion(0, 0)

# Funciones auxiliares para obtener los primeros n numeros de cada tipo
def primeros_n_numeros(funcion, cantidad):
    # Retorna una lista con los primeros 'cantidad' numeros que cumplen la funcion booleana
    encontrados = []
    numero = 1
    while len(encontrados) < cantidad:
        if funcion(numero):
            encontrados.append(numero)
        numero += 1
    return encontrados

# Ejemplo de uso para imprimir los primeros 30 de cada tipo
print("Primeros 30 numeros defectivos:", primeros_n_numeros(numeros_defectivos, 30))
print("Primeros 30 numeros abundantes:", primeros_n_numeros(numeros_abundantes, 30))
print("Primeros 30 numeros semiperfectos:", primeros_n_numeros(numeros_semiperfectos, 30))
print("Primeros 30 numeros perfectos:", primeros_n_numeros(numeros_perfectos, 30))
print("Primeros 30 numeros primos:", primeros_n_numeros(numeros_primos, 30))
````

Una característica útil de las funciones definidas por el usuario es la **recursion**, la capacidad de una función de llamarse a sí misma. Por ejemplo, considere la siguiente definición del factorial $n!$ de un entero positivo $n$:

$$n! = 
\begin{cases} 
1 & \text{si } n = 1, \\ 
n \times (n - 1)! & \text{si } n > 1.
\end{cases}$$

Esta constituye una definición completa del factorial que nos permite calcular el valor de $n!$ para cualquier entero positivo. Podemos emplear esta definición directamente para crear una función en Python para factoriales, así:



```python
def factorial(n):
    if n == 1:
        return 1
    else:
        return n * factorial(n-1)
```
- Los números de Catalan $C_n$ aparecieron previamente en clase. La definición dada allí puede reescribirse como:

$$C_n =

1 & \text{si } n = 0, \\

Escriba una función en Python, usando recursión, que calcule $C_n$. Utilice su función para calcular e imprimir $C_{50}$.

- Euclides demostró que el máximo común divisor $g(m, n)$ de dos enteros no negativos $m$ y $n$ satisface:

$$g(m, n) =
\begin{cases} 
m & \text{si } n = 0, \\
g(n, m \bmod n) & \text{si } n > 0.
\end{cases}$$

Escriba una función en Python $g(m, n)$ que emplee recursión para calcular el máximo común divisor de $m$ y $n$ usando esta fórmula. Use su función para calcular e imprimir el máximo común divisor de 108 y 192.


````python
# Funcion recursiva para calcular el factorial de un numero entero positivo
def factorial(n):
    # Si el numero es 1, el factorial es 1 (caso base)
    if n == 1:
        return 1
    else:
        # Si n > 1, se multiplica por el factorial del anterior
        return n * factorial(n - 1)

# Funcion recursiva para calcular el enesimo numero de Catalan
def numero_catalan(n):
    # Si n es 0, el numero de Catalan es 1 (caso base)
    if n == 0:
        return 1
    else:
        # Usamos la formula recursiva de Catalan
        return ((4 * n - 2) * numero_catalan(n - 1)) // (n + 1)

# Funcion recursiva para calcular el maximo comun divisor (mcd) usando el algoritmo de Euclides
def mcd(m, n):
    # Si n es 0, el mcd es m (caso base)
    if n == 0:
        return m
    else:
        # Llamada recursiva con n y el residuo de m entre n
        return mcd(n, m % n)

# Calcular y mostrar el numero de Catalan C_50
c50 = numero_catalan(50)
print("El numero de Catalan C_50 es:", c50)

# Calcular y mostrar el mcd de 108 y 192
resultado_mcd = mcd(108, 192)
print("El maximo comun divisor de 108 y 192 es:", resultado_mcd)
````

Resolviendo la ecuación de Schrödinger, se puede demostrar que cuando $E > V$ la partícula puede:
- Pasar el escalón, en cuyo caso tendrá una energía cinética menor $E - V$ al otro lado y un vector de onda correspondientemente más pequeño $k_2 = \sqrt{2m(E - V)}/\hbar$
- Ser reflejada, manteniendo toda su energía cinética y vector de onda sin cambios, pero moviéndose en dirección opuesta.


Las probabilidades $T$ y $R$ para transmisión y reflexión están dadas por:

$$T = \frac{4k_1k_2}{(k_1 + k_2)^2}, \quad R = \left(\frac{k_1 - k_2}{k_1 + k_2}\right)^2.$$

Consideremos una partícula con masa igual a la del electrón $m = 9.11 \times 10^{-31} \, \text{kg}$ y energía $10 \, \text{eV}$ que encuentra un escalón de potencial de altura $9 \, \text{eV}$. La órbita en el espacio de un cuerpo alrededor de otro, como un planeta alrededor del Sol, no necesita ser circular. En general toma la forma de una elipse, con el cuerpo a veces más cerca y a veces más lejos. Si se conocen la distancia $\ell_{1}$ del acercamiento más próximo que hace un planeta al Sol, llamado perihelio, y su velocidad lineal $v_{1}$ en el perihelio, entonces cualquier otra propiedad de la órbita puede calcularse a partir de estos dos datos de la siguiente manera.
La segunda ley de Kepler nos dice que la distancia $\ell_{2}$ y velocidad $v_{2}$ del planeta en su punto más distante, o afelio, satisfacen $\ell_{2}v_{2}=\ell_{1}v_{1}$. Al mismo tiempo, la energía total, cinética más gravitacional, de un planeta con velocidad $v$ y distancia $r$ del Sol está dada por
    $$
    E = \tfrac{1}{2}mv^{2} - G\frac{mM}{r},
    $$
    donde $m$ es la masa del planeta, $M=1.9891\times 10^{30}\,\mathrm{kg}$ es la masa del Sol, y $G=6.6738\times 10^{-11}\,\mathrm{m}^{3}\,\mathrm{kg}^{-1}\,\mathrm{s}^{-2}$ es la constante gravitacional de Newton. Dado que la energía debe conservarse, se llega a que $v_{2}$ es la raíz más pequeña de la ecuación cuadrática
    $$
    v_{2}^{2} - \frac{2GM}{v_{1}\ell_{1}}v_{2} - \left[v_{1}^{2} - \frac{2GM}{\ell_{1}}\right] = 0.
    $$
    Una vez que se tiene $v_{2}$, podemos calcular $\ell_{2}$ usando la relación $\ell_{2}=\ell_{1}v_{1}/v_{2}$.

- Dados los valores de $v_{1}$, $\ell_{1}$, y $\ell_{2}$, otros parámetros de la órbita vienen dados por fórmulas simples que pueden derivarse de las leyes de Kepler y del hecho de que la órbita es una elipse:
$$
    \text{Semieje mayor:}  \quad a = \tfrac{1}{2}(\ell_{1}+\ell_{2}), 
$$
$$
    \text{Semieje menor:}  \quad b = \sqrt{\ell_{1}\ell_{2}}, 
$$
$$
    \text{Periodo orbital:}  \quad T = \frac{2\pi ab}{\ell_{1}v_{1}}, 
$$
$$
    \text{Excentricidad orbital:}  \quad e = \frac{\ell_{2}-\ell_{1}}{\ell_{2}+\ell_{1}}.
$$

Escriba un programa que solicite al usuario ingresar la distancia al Sol y la velocidad en el perihelio, luego calcule e imprima las cantidades $\ell_{2}$, $v_{2}$, $T$, y $e$.

- Pruebe su programa calculando las propiedades de las órbitas de la Tierra (para la cual $\ell_{1}=1.4710\times 10^{11}\,\mathrm{m}$ y $v_{1}=3.0287\times 10^{4}\,\mathrm{m}\,\mathrm{s}^{-1}$) y del cometa Halley ($\ell_{1}=8.7830\times 10^{10}\,\mathrm{m}$ y $v_{1}=5.4529\times 10^{4}\,\mathrm{m}\,\mathrm{s}^{-1}$). Entre otras cosas, debería encontrar que el período orbital de la Tierra es un año y el del cometa Halley es alrededor de 76 años.


````python
# Constantes físicas
G = 6.6738e-11  # m^3 kg^-1 s^-2
M = 1.9891e30   # masa del Sol en kg

def calcular_orbita(l1, v1):
    """
    Calcula los parámetros orbitales a partir del perihelio (l1, v1).
    Retorna: l2, v2, periodo T (en años), excentricidad e.
    """
    A = 1
    B = -2 * G * M / (v1 * l1)
    C = -(v1**2 - 2 * G * M / l1)

    discriminante = B**2 - 4 * A * C
    v2a = (-B + np.sqrt(discriminante)) / (2 * A)
    v2b = (-B - np.sqrt(discriminante)) / (2 * A)
    v2 = min(v2a, v2b)

    l2 = l1 * v1 / v2
    a = 0.5 * (l1 + l2)
    b = np.sqrt(l1 * l2)
    T = 2 * np.pi * a * b / (l1 * v1)
    T_anios = T / (60 * 60 * 24 * 365.25)
    e = (l2 - l1) / (l2 + l1)

    return l2, v2, T_anios, e

# ------------------------------
# PRUEBA: Datos para la Tierra
l1_tierra = 1.4710e11     # m
v1_tierra = 3.0287e4      # m/s

l2, v2, T, e = calcular_orbita(l1_tierra, v1_tierra)

print("Resultados para la Tierra:")
print("l2 =", l2, "m")
print("v2 =", v2, "m/s")
print("Periodo orbital =", T, "años")
print("Excentricidad =", e)

# ------------------------------
# PRUEBA: Datos para el cometa Halley
l1_halley = 8.7830e10     # m
v1_halley = 5.4529e4      # m/s

l2, v2, T, e = calcular_orbita(l1_halley, v1_halley)

print("\nResultados para el cometa Halley:")
print("l2 =", l2, "m")
print("v2 =", v2, "m/s")
print("Periodo orbital =", T, "años")
print("Excentricidad =", e)
````

# 6. 
Un proyectil es lanzado desde una altura inicial $y_0 = 0$ con una velocidad inicial de $4.0\, \text{m/s}$ a $45^\circ$ respecto a la horizontal. Graficar la trayectoria del proyectil en su plano de movimiento (considerando el eje $x$ paralelo a la horizontal y el origen en la posición inicial del proyectil) desde el punto de lanzamiento hasta el punto donde impacta el suelo (asumiendo un terreno plano). La gráfica debe incluir etiquetas en los ejes y un título. Recuerde que las ecuaciones del movimiento parabólico están dadas por

$$
\text{Posición horizontal:} \quad x(t) = v_{0x} t = v_0 \cos\theta \cdot t 
$$
$$
\text{Posición vertical:} \quad y(t) = y_0 + v_{0y} t - \frac{1}{2}gt^2 = y_0 + v_0 \sin\theta \cdot t - \frac{1}{2}gt^2 
$$
$$
\text{Velocidad horizontal:} \quad v_x(t) = v_{0x} =v_0 \cos \theta= \text{constante} 
$$
$$
\text{Velocidad vertical:} \quad v_y(t) = v_{0y} - gt = v_0 \sin\theta - gt
$$

````python
# funcion que calcula y grafica la trayectoria de un proyectil
# entradas:
#    v0: velocidad inicial en m/s (float)
#    angulo_grados: angulo de lanzamiento en grados (float)
#    y0: altura inicial en metros (float)
# salida:
#    grafica de x(t) vs y(t) desde el lanzamiento hasta el impacto
def graficar_trayectoria_proyectil(v0, angulo_grados, y0):
    # aceleracion de la gravedad (m/s^2)
    g = 9.81

    # convertir angulo a radianes
    angulo_rad = np.radians(angulo_grados)

    # componentes de la velocidad inicial
    v0x = v0 * np.cos(angulo_rad)
    v0y = v0 * np.sin(angulo_rad)

    # coeficientes de la ecuacion cuadratica para y(t)=0: A*t^2 + B*t + C = 0
    # donde y(t) = y0 + v0y*t - 0.5*g*t^2.
    # Por lo tanto: A = -0.5*g, B = v0y, C = y0.
    # (En el codigo original, 'a' hace referencia a A, 'b' a B y 'c' a C)
    coeff_A = -0.5 * g
    coeff_B = v0y
    coeff_C = y0

    # discriminante de la ecuacion cuadratica B^2 - 4AC
    discriminante = coeff_B**2 - 4 * coeff_A * coeff_C

    # tiempo de impacto (raiz positiva de la ecuacion cuadratica que tiene sentido fisico)
    # Las raices son t = (-B +/- sqrt(discriminante)) / (2A)
    # La raiz (-B - sqrt(discriminante)) / (2A) es la que consistentemente da el tiempo de vuelo positivo.
    if discriminante < 0:
        # Esto no deberia ocurrir si y0 >= 0 (proyectil lanzado desde el suelo o arriba)
        # porque v0y^2 + 2*g*y0 siempre seria no negativo.
        # Si ocurriera, significa que el proyectil nunca alcanza y=0 en este modelo.
        print("Advertencia: El discriminante es negativo. El proyectil podría no alcanzar y=0.")
        t_impacto = 0 # No hay impacto con y=0 o es un caso no contemplado.
    else:
        t_impacto = (-coeff_B - np.sqrt(discriminante)) / (2 * coeff_A)

    # Asegurar que el tiempo de impacto no sea negativo (p.ej. si y0<0 y se lanza hacia arriba)
    # Para y0>=0 y la raiz correcta, t_impacto deberia ser >=0.
    t_final_simulacion = max(0, t_impacto)

    # vector de tiempos desde 0 hasta t_impacto
    # Si t_final_simulacion es 0, np.linspace(0,0,N) produce un array de ceros, lo cual es correcto
    # para graficar un punto si no hay movimiento.
    t = np.linspace(0, t_final_simulacion, 500)

    # posiciones en cada instante
    x = v0x * t
    y = y0 + v0y * t + coeff_A * t**2 # coeff_A es -0.5*g

    # graficar trayectoria
    plt.figure(figsize=(8, 5))
    plt.plot(x, y, label='Trayectoria del proyectil', color='teal') # Etiqueta más genérica
    plt.title('Trayectoria de un Proyectil') # Título más genérico
    plt.xlabel('Posición horizontal x (m)')
    plt.ylabel('Posición vertical y (m)')
    plt.axhline(0, color='gray', linestyle='--', linewidth=0.7) # Linea de referencia para el suelo
    plt.grid(True)
    plt.legend()
    plt.show()

# invocar la funcion con los valores dados
graficar_trayectoria_proyectil(4.0, 45, 0.0)

````

Los polinomios de Legendre son de gran importancia en física, especialmente en la solución de la ecuación de Laplace en coordenadas esféricas. Son fundamentales en la expansión en armónicos esféricos, lo que los hace esenciales en problemas de mecánica cuántica y electromagnetismo. Cada polinomio de Legendre $P_n(x)$ es un polinomio de grado $n$. Este puede ser expresado usando la Fórmula de Rodrigues:
     
$$
P_n (x)=\frac{1}{n!\ 2^n} \frac{d^n}{dx^n} (x^2-1)^n, \ \ n \in  \mathbb{Z^{+}}+\{0\}
$$
        
- Utilizando la fórmula de Rodrigues halle $P_0(x)$ y $P_1(x)$.
-  Con la fórmula de recurrencia
$$
            P_{j+1}(x)=\frac{(2j+1)x P_j (x)-jP_{j-1}(x)}{j+1},
$$
elabore un código que permita calcular $P_n(x)$ para $n\geq2$.
- En una misma gráfica muestre $P_n(x)$ para $1\leq n\leq5$.


````python
# funcion que aplica la formula de Rodrigues para P0(x) y P1(x)
# entradas:
#   x_vals: array de valores de x donde evaluar (numpy array)
# salidas:
#   P0_vals: array con P0(x) para cada x
#   P1_vals: array con P1(x) para cada x
def legendre_rodrigues_base(x_vals):
    # P0(x) = 1 segun formula de Rodrigues para n=0
    P0_vals = np.ones_like(x_vals)
    # P1(x) = x segun formula de Rodrigues para n=1
    P1_vals = x_vals.copy()
    return P0_vals, P1_vals

# funcion que calcula Pn(x) usando la relacion de recurrencia para n>=2
# entradas:
#   n: grado del polinomio (int >= 0)
#   x_vals: array de valores de x donde evaluar (numpy array)
# salida:
#   Pn_vals: array con Pn(x) para cada x en x_vals
def legendre_recurrence(n, x_vals):
    # casos base n=0 y n=1
    if n == 0:
        return np.ones_like(x_vals)
    if n == 1:
        return x_vals.copy()
    # inicializar P_{n-2} y P_{n-1} con P0 y P1
    P_nm2, P_nm1 = legendre_rodrigues_base(x_vals)
    # aplicar la formula de recurrencia hasta grado n
    for j in range(1, n):
        Pn_vals = ((2*j + 1) * x_vals * P_nm1 - j * P_nm2) / (j + 1)
        P_nm2, P_nm1 = P_nm1, Pn_vals
    return Pn_vals

# generar valores de x en el intervalo [-1, 1]
x_vals = np.linspace(-1, 1, 400)

# calcular P0 y P1 via Rodrigues
P0_vals, P1_vals = legendre_rodrigues_base(x_vals)

# mostrar en consola un valor de ejemplo para P0 y P1
print("Ejemplo: P0(-1) =", P0_vals[0], ", P1(-1) =", P1_vals[0])

# graficar Pn(x) para n = 1..5
plt.figure(figsize=(8, 5))
for n in range(1, 6):
    y_vals = legendre_recurrence(n, x_vals)
    plt.plot(x_vals, y_vals, label=f'P_{n}(x)')

plt.title('Polinomios de Legendre P_n(x) para n = 1 a 5')
plt.xlabel('x')
plt.ylabel('P_n(x)')
plt.grid(True)
plt.legend()
plt.show()
````
