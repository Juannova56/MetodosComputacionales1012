```python
import numpy as np
import  matplotlib.pyplot as plt
import math
```

```python
#Punto 1
def divisores_propios(n):
  # Esta funcion retorna la lista de los divisores de n, salvo el mismo. se recorren todos los numeros del 1 hasta n-1 y se agregan a lista los divisores de n
  return [x for x in range(1,n) if n % x == 0]

def numeros_defectivos(a):
  divisores = divisores_propios(a)
  b = sum(divisores)
  if b < a:
    return True
  else:
    return False

def numeros_abundantes(a):
  divisores = divisores_propios(a)
  b = sum(divisores)
  if b > a:
    return True
  else:
    return False

def numeros_perfectos(a):
  divisores = divisores_propios(a)
  b = sum(divisores)
  if b==a:
    return True
  else:
    return False

def numeros_primos(a):
  divisores = divisores_propios(a)
  b = sum(divisores)
  if b == 1:
    return True
  else:
    return False
```

```python
#Punto 2
def es_palindromo(n):
    n_str = str(n) #Convertimos el numero a un string
    return n_str == n_str[::-1] #Ya una vez lo tengamos como string, lo comparamos con su inverso

numero = int(input("Ingrese un número entero: "))

if es_palindromo(numero): #Usando la función ya dependiendo si nos bota un true o false, nos dira si es o no un numero palindromo
    print("Es un palíndromo")
else:
    print("No es un palíndromo")
```

```python
#Punto 3
def catalan(n):
    if n == 0:
        return 1
    else:
        return (4 * n - 2) * catalan(n - 1) // (n + 1) #Se usa la iteración tal y como nos lo piden y la división entera la asegurarnos que nos bote un entero y no un float.

print("C_50 =", catalan(50))

def mcd(m, n):
    if n == 0:
        return m
    else:
        return mcd(n, m % n)

print("MCD(108, 192) =", mcd(108, 192))
```

```python
#Punto 4
h = 1.0545718e-34  #Constante de Planck en J*s
eV = 1.60218e-19 #1 eV en joules
m = 9.11e-31 #Masa del electrón en kg
#Energía cinetica y potencial en eV
E_eV = 10
V_eV = 9
# Convertimos a joules
E = E_eV * eV
V = V_eV * eV

k1 = (2*m*E)**(1/2) / h #Calculamos los vectores de onda k1 y k2
k2 = (2*m*(E-V))**(1/2) / h

T = (4 * k1 * k2) / ((k1 + k2)**2) #Probabilidades de transmisión y reflexión
R = ((k1 - k2) / (k1 + k2))**2

print(f"Probabilidad de transmisión T = {T:.4f}")
print(f"Probabilidad de reflexión R = {R:.4f}")
print(f"(T + R) = {T + R:.4f}") #Se verifica que efectivamente la suma de estas dos probabilidades sea igual al 100%
```

```python
#Punto 5
M = 1.9891e30
pi = math.pi
G = 6.6743e-11
def orb(l1, v1):
    v2 = ((2*G*M)/(v1*l1)+(-v1**2+(2*G*M)/l1))**(1/2) #Se despeja la primera formula
    l2 = (l1*v1)/v2
    e = (l2-l1)/(l2+l1)
    a = (l1+l2)/2
    b = (l1*l2)**(1/2)
    T = (2*pi*a*b)/(l1*v1)
    return l2, v2, T, e

#Propiedades iniciales de los astros
# para tierra
l1T = 1.4710e11
v1T = 3.0287e4
#para halley
l1H = 8.7830e10
v1H = 5.4529e4
# Diccionario con el astro y sus propiedades
datos = {
    'Tierra' : orb(l1T, v1T),
    'Cometa Halley' : orb(l1H, v1H)
}
for astro in datos.keys():
    print(f'Los datos orbitales del afelio, velocidad en el afelio, periodo orbital y exentricidad de {astro} son respectivamente:\n{datos[astro]}\n')
```

```python
#Punto 6

def graf(y0, v0, ang): #Se pide para la función la altura inicial, la velocidad e inicial y también el angulo con el que se tira en grados
  ang1 = ang*np.pi/180 #Conversión de los grados a radianes
  g = 9.8
  v0x = v0*np.cos(ang1) #Descomposición en x de la velocidad inicial
  v0y = v0*np.sin(ang1) #Descomposión en y de la velocidad inicial
  t_f = (v0y+((v0y**2)+(2*g*y0))**(0.5))/g #Se calcula el tiempo total el cual el proyectil va a necesitar para llegar al suelo
  x_f = v0x*t_f #La distancia total recorrida en el eje x
  y_f = 0
  ti = np.linspace(0, t_f, 1000)
  yi = y0 + v0y*ti - (1/2)*g*(ti**2) #Posiciones en y a lo largo de la trayectoria del proyectil
  xi = v0x*ti #Posiciones en x a lo largo de la trayectoria del proyectil

  fig, ax=plt.subplots()
  ax.plot(xi, yi) #Como nos piden un gráfico de posición en x vs posición en y, se grafican tal cual

  return plt.show
```

```python
#Punto 7
import sympy as sp
# variable simbólica para imprimir el polinomio sin evaluarlo
x = sp.symbols('x')
def Formula_Rodrigues(n , df_n):
  coeficiente = 1 / (math.factorial(n) * 2**n)
  return coeficiente * df_n
n = 0
f_0 = 1 # (x^2 - 1)^n cuando n=0
P_0 = Formula_Rodrigues(n , 1)

print(f'P_{n}(x) =', P_0)

n = 1
f_1 = x**2 - 1 # (x^2 - 1)^n cuando n=1
df_1 = 2 * x   # derivada de f_1
P_1 = Formula_Rodrigues(n , df_1)

print(f'P_{n}(x) =', P_1)

def P_Legendre(n):
 # Utilizamos la informacion anterior para poder realizar las iteraciones necesarias para llegar a los n pedidos
  if n==0:
    return 1
  elif n==1:
    return x
  else:
    # Comenzamos con los valores iniciales de P_j y P_(j-1)
    Pj= x
    Pj_anterior= 1
    for j in range(1 , n):

      # En cuentro P_(j+1) usando la fórmula de recurrencia
      Pj_siguiente = ((2*j + 1) * x * Pj - j * Pj_anterior)/(j + 1)

      # Guardo los polinomios P_(j-1) y P_(j) para la siguiente iteracion
      Pj_anterior = Pj
      Pj = Pj_siguiente

    return Pj

P_2 = P_Legendre(3)
print(f'P_{2}(x) = {P_2}')
P_3 = P_Legendre(4)
print(f'P_{3}(x) = {P_3}')
P_4 = P_Legendre(5)
print(f'P_{4}(x) = {P_4}')
P_5 = P_Legendre(6)
print(f'P_{5}(x) = {P_5}')

x_vals = np.linspace(-1, 1, 1000) #Valores entre -1 y 1 para poder ver bien las funciones graficadas
for n in range(1, 6):
    Pn_expresion = P_Legendre(n) # Esto nos da una exprecion con valores simbolicos, por tanto, no se puede evaluar
    Pn_funcion = sp.lambdify(x, Pn_expresion, modules='numpy') # Convertimos la expresion a una que podamos evaluar
    y_vals = Pn_funcion(x_vals)
    plt.plot(x_vals, y_vals, label=f'P_{n}(x)')

plt.title('Polinomios de Legendre P_n(x) para 1 <= n <= 5')
plt.grid()
plt.legend()
plt.show()
```
