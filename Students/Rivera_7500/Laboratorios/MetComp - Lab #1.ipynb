```python
#Problema 1
import itertools as it

print("Vamos a categorizar tu numero natural por medio de unos filtros")
x = 12 #<--- Este es tu número NATURAL

"""
Divisores de nuestro numero
"""


def Div(x):
    Divisores = []
    for i in range(1,x+1):
        if x%i == 0:
            Divisores.append(i)
        else:
            pass
    return Divisores

"""
Numeros defectivos
"""

def numeros_defectivos(x):
    if sum(Div(x)) - x < x:
        return True
    else:
        return False

"""
Numeros abundandes
"""

def numeros_abundantes(x):
    if sum(Div(x)) - x > x:
        return True
    else:
        return False
    



#---------------------------------------------------------------------------------

"""
Numeros semiperfectos
"""
def numeros_semiperfectos(x):
    for i in range(1,len(Div(x)[0:len(Div(x))-1])+1):
        for n in it.combinations(Div(x)[0:len(Div(x))-1],i):
            if sum(n) == x:
                return True
            else:
                pass
    return False



#---------------------------------------------------------------------------------
"""
Numeros perfectos
"""

def numeros_perfectos(x):
    if sum(Div(x)) - x == x:
        return True
    else:
        return False

"""
Numeros primos
"""

def numeros_primos(x):
    if len(Div(x)) == 2:
        return True
    else:
        return False

print("Tu numero seleccionado --->",x,"\n",
"\nNumero defectivo = ",numeros_defectivos(x),
"\nNumero abundante = ",numeros_abundantes(x),
"\nNumero semiperfecto = ",numeros_semiperfectos(x),
"\nNumero perfecto = ",numeros_perfectos(x),
"\nNumero primo = ",numeros_primos(x))

print("\n---------------------Ejercicio extra---------------------\n")

nu=[]
n = 0
while len(nu) < 30:
    n += 1
    if numeros_defectivos(n) == True:
        nu.append(n)
    else:
        pass

print(" Los primeros 30 numeros defectivos son:\n",nu)

nu=[]
n = 0
while len(nu) < 30:
    n += 1
    if numeros_abundantes(n) == True:
        nu.append(n)
    else:
        pass

print(" Los primeros 30 numeros abundantes son:\n",nu)

nu=[]
n = 0
while len(nu) < 30:
    n += 1
    if numeros_semiperfectos(n) == True:
        nu.append(n)
    else:
        pass

print(" Los primeros 30 numeros semiperfectos son:\n",nu)

nu=[]
n = 0
while len(nu) < 4:
    n += 1
    if numeros_perfectos(n) == True:
        nu.append(n)
    else:
        pass

print(" Los primeros 4 numeros perfectos son(Limitado por el bienestar de la maquina):\n",nu)

nu=[]
n = 0
while len(nu) < 30:
    n += 1
    if numeros_primos(n) == True:
        nu.append(n)
    else:
        pass

print(" Los primeros 30 numeros primos son:\n",nu)


```

```python
#Problema 5
import numpy as np

def plan(l1,v1):
    M = 1.9891e30  #<--- Cte numero 1
    G = 6.6738e-11   #<--- Cte numero 2
    
    a = 1
    b = -( (2*G*M) / (v1*l1) )
    c = -((v1**2) - ( (2*G*M) / l1))

    r1 = ( -b + (b**2 - 4*a*c) ** (1/2) ) / 2*a
    r2 = ( -b - (b**2 - 4*a*c) ** (1/2) ) / 2*a

    if r1 <= r2:
        v2 = r1
    else:
        v2 = r2

    l2 = l1*v1 / v2

    Sma = (1/2)*(l1+l2)
    Sme = (l1*l2)**(1/2)
    T = (2*np.pi*Sma*Sme) / (l1*v1)
    e = (l2-l1) / (l2+l1)

    return (v2,l2,T,e)

"""
Propiedades de la orbita de la tierra
"""

l1 = 1.4710e11   #<--- Dato numero 1
v1 = 3.0287e4   #<--- Dato numero 2

v2,l2,T,e = plan(l1,v1)
print("-----Siguiendo que tus datos iniciales fueron (TIERRA)-----\n",
"\nVelocidad 1(m/s)  = ",v1,
"\nLongitud 1(m) = ",l1,
"\n\nEntonces podemos calcular la sus segundas componentes:",
"\nVelocidad 2(m/s)  = ",v2,
"\nLongitud 2(m) = ",l2,
"\n\nPor lo que podemos determinar que su periodo y su excentricidad son:",
"\nPeriodo(s) = ",T,"Periodo(Años) = ",T*3.171e-8,
"\nExcentricidad = ",e,)

print("\n_____Confirmando el periodo de aproximadamente un año_____")

"""
Cometa Halley
"""

l1 = 8.7830e10   #<--- Dato numero 1
v1 = 5.4529e4   #<--- Dato numero 2

v2,l2,T,e = plan(l1,v1)
print("\n-----S(iguiendo que tus datos iniciales fueron (COMETA HALLEY)-----\n",
"\nVelocidad 1(m/s) = ",v1,
"\nLongitud 1(m) = ",l1,
"\n\nEntonces podemos calcular la sus segundas componentes:",
"\nVelocidad 2(m/s)  = ",v2,
"\nLongitud 2(m) = ",l2,
"\n\nPor lo que podemos determinar que su periodo y su excentricidad son:",
"\nPeriodo(s) = ",T,"Periodo(Años) = ",T*3.171e-8,
"\nExcentricidad = ",e,)

print("\n_____Confirmando el periodo de aproximadamente 76 años_____")
```
