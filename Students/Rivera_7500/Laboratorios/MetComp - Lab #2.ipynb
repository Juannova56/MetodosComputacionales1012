```python
#Problema 1

Num = 1000011100001    #<--------- Tu numero

print("Vamos a convertir tu numero binario de 16 BITS a base decimal:")
print("Tu numero a convertir sera:",Num)

def myint(x):
    x = list(reversed(str(x)))
    if len(x) > 16:
        raise RuntimeError("Tu numero no es de 16 bits")
    else:pass
    N = 0 
    for i in range(0,len(x)):
        N += (int(x[i])) * (2**i)
    return N

print("Tu numero convertido a base decimal es:",myint(Num),"\n")
print("Segun la comprobacion con la funcion preestablecida estamos:",int(str(Num),2) == myint(Num),"\nPor lo que estamos en lo correcto :)")

```

```python
#Problema 3
import numpy as np

print()
print("---Metodo básico 101---")
print()

def pi(it , tol , x = 1 ):
    ndei = 0
    arctanx = 0
    for i in range(1,it+1):
        ndei +=1
        n = ( ((-1)**(i+1)) * (x**(2*i-1)) ) / (2*i -1)
        arctanx += n
        pi = 4*arctanx
        if abs(pi - np.pi) < tol:
            return pi , ndei
        else:
            pass

pi_e , n_de_i = pi(1001 , 1e-3) 
print(" Nuestro valor para pi es de:",pi_e,"\n Para obtener una tolerancia de 1e-3 se necesitaron este numero de terminos:",n_de_i)
print()
pi_e , n_de_i = pi(10001 , 1e-4) 
print(" Nuestro valor para pi es de:",pi_e,"\n Para obtener una tolerancia de 1e-4 se necesitaron este numero de terminos:",n_de_i)

#--------------------------------------------------------

print()
print("---Metodo avanzado B)---")
print()

def pimejor(it , tol ):
    ndei = 0
    arctan12 = 0
    arctan13 = 0

    for i in range(1,it+1):
        ndei +=1
        n = ( ((-1)**(i+1)) * ((1/2)**(2*i-1)) ) / (2*i -1)
        arctan12 += n

        n = ( ((-1)**(i+1)) * ((1/3)**(2*i-1)) ) / (2*i -1)
        arctan13 += n

        pi = 4*(arctan12 + arctan13)

        if abs(pi - np.pi) < tol:
            return pi , ndei
        else:
            pass

pi_e , n_de_i = pimejor(10 , 1e-3) 
print(" Nuestro valor para pi (Mejor) es de:",pi_e,"\n Para obtener una tolerancia de 1e-3 se necesitaron este numero de terminos:",n_de_i)
print()
pi_e , n_de_i = pimejor(100 , 1e-6) 
print(" Nuestro valor para pi (Mejor) es de:",pi_e,"\n Para obtener una tolerancia de 1e-6 se necesitaron este numero de terminos:",n_de_i)

#--------------------------------------------------------

print()
print("---Metodo profesional =0---")
print()

def pipreciso(it , tol ):
    ndei = 0
    arctan15 = 0
    arctan1239 = 0

    for i in range(1,it+1):
        ndei +=1
        n = ( ((-1)**(i+1)) * ((1/5)**(2*i-1)) ) / (2*i -1)
        arctan15 += n

        n = ( ((-1)**(i+1)) * ((1/239)**(2*i-1)) ) / (2*i -1)
        arctan1239 += n

        pi = 4*(4*arctan15 - arctan1239)

        if abs(pi - np.pi) < tol:
            return pi , ndei
        else:
            pass

pi_e , n_de_i = pipreciso(10 , 1e-3) 
print(" Nuestro valor para pi (Preciso) es de:",pi_e,"\n Para obtener una tolerancia de 1e-3 se necesitaron este numero de terminos:",n_de_i)
print()
pi_e , n_de_i = pipreciso(10 , 1e-6) 
print(" Nuestro valor para pi (Preciso) es de:",pi_e,"\n Para obtener una tolerancia de 1e-6 se necesitaron este numero de terminos:",n_de_i)

print("\n-Podemos percatarnos que el metodo basico consume muchos terminos\n-El avanzado lo reduce increiblemente demasiado \n-El profesional lo reduce muy poco pero reduce")
```

```python
#Problema 4
import math as m

def exp_x(x,it,tol):
    mi = []
    exp_1 = 0
    for i in range(0,it):
        n = (((-1)**i) * (x**i)) / (m.factorial(i))
        mi.append(n)
        exp_1 += n 
        if abs( ( (-x)**(i+1) ) / ( m.factorial(i+1) ) ) <= tol*exp_1:
            return exp_1 , mi
        else:
            pass

for n in [1,0.75,0.5,0.25]:
    exp , iter = exp_x(n,1000,1e-7)
    print("\nPara x =", n ,"la serie converge a =",exp,"\nTomandole el siguiente numero de iteraciones = ",len(iter))
    print("Tiene un error porcentual de", (abs(np.exp(-n) - exp) / abs(np.exp(-n)))*100,"%")

print("\n------Entre mas nos alejamos de 1 su error disminuye--------")

for n in [9,10,11]:
    exp , iter = exp_x(n,1000,1e-7)
    print("\nPara x =", n ,"la serie converge a =",exp,"\nTomandole el siguiente numero de iteraciones = ",len(iter))
    print("Tiene un error porcentual de", (abs(np.exp(-n) - exp) / abs(np.exp(-n)))*100,"%")

print("\n---------Su cancelación se puede evidenciar cuando vemos sus terminos---------\n"
,exp_x(10,1000,1e-7)[1][9],"+",exp_x(10,1000,1e-7)[1][9] ,"=",exp_x(10,1000,1e-7)[1][9]+exp_x(10,1000,1e-7)[1][10])
print("------------Donde se pierde mucha presición------------")

def expx(x,it,tol):
    mi = []
    exp_1 = 0
    for i in range(0,it):
        n = ((x**i)) / (m.factorial(i))
        mi.append(n)
        exp_1 += n 
        if abs( ( (x)**(i+1) ) / ( m.factorial(i+1) ) ) <= tol*exp_1:
            return exp_1 , mi
        else:
            pass

for n in [100,200]:
    exp , iter = expx(n,1000,1e-7)
    print("\nPara x =", n ,"la serie converge a =",1 / exp,"\nTomandole el siguiente numero de iteraciones = ",len(iter))
    print("Tiene un error porcentual de", (abs((1/np.exp(n)) - (1/exp)) / abs(1/np.exp(n)))*100,"%")

print("\n------Para terminos muy altos esta modificación nos ayuda eliminando la cancelación susbstractiva--------")

N = []
for n in range(10,22):
    exp , iter = exp_x(n,1000,1e-7)
    print("\nPara x =", n ,"la serie converge a =",exp,"\nTomandole el siguiente numero de iteraciones = ",len(iter))
    print("Tiene un error porcentual de", (abs(np.exp(-n) - exp) / abs(np.exp(-n)))*100,"%")
    N.append((abs(np.exp(-n) - exp) / abs(np.exp(-n)))*100)

print("\n------A partir de x = 19 la serie empieza a perder gran precisión para finalmente divergir en X=20--------")

print("\nEn el siguiente grafico nos percatamos como diverge cuando alcanza aproximadamente el valor de 20")
plt.style.use("classic")
plt.title("Errores en la serie")
plt.xlabel("Preimagenes")
plt.ylabel("Errores porcentuales (%)")
plt.plot(range(10,22) , N)
plt.show()
```

```python
#Problema 5

x = 1   #<-------- Valor para evaluar la derivada
def funcion(x):
    return x*(x-1)

print("El valor de nuestra funcion en x =",x,"es de:",funcion(x))

def derivada(x,s,f = funcion):
    return ( f(x+s) - f(x) ) / (s)

print("El valor de nuestra derivada con s=1e-2 en x =",x,"es de:",derivada(x,1e-2))
print("\nDe forma analita nuestra derivada es de 2x-1 por lo que en X=1 es de:",1)

print("\nNunca sera 1 porque debemos que hacer un cambio infinitamente pequeño lo\nque es imposible para nuestra maquina, a no ser de que redondee y obtenga el valor.\n")

for i in [1e-4,1e-6,1e-8,1e-10,1e-12,1e-14]:
    print("El valor de nuestra derivada con s=",i," en x =",x,"es de:",derivada(x,i))

print("\nEl comportamiento se debe a la falta de presición de los ultimos decimales,\nal hacer un cambio mas pequeño la maquina no podra mantener tanta presición y empeorara.")
```
