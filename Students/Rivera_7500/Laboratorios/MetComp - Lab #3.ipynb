```python
#-----Problema 1-----

#---------------------------------------------------------------------------------------------------------------------------

""" ----------------------------------------- Preguntas a contestar ----------------------------------------- """

#---------------------------------------------------------------------------------------------------------------------------

print("1.) Â¿Cual es la anomalia excentrica en abril 1 de 1968?")
print("2.) Grafica de la excentricidad y el periodo completo de el cometa")


#---------------------------------------------------------------------------------------------------------------------------

""" ----------------------------------------- Librerias ----------------------------------------- """

#---------------------------------------------------------------------------------------------------------------------------

import numpy as np
import matplotlib.pyplot as plt
import scipy as sp




#---------------------------------------------------------------------------------------------------------------------------

""" ----------------------------------------- Varaibles y paremetros usados ----------------------------------------- """

#---------------------------------------------------------------------------------------------------------------------------

e = 0.9672671                # Excentricidad
T = 75.96000                 # Periodo orbital
to = 1986.1113               # Febrero 9 de 1968
t = 1986 + (91/365)          # Abril 1 de 1968

E = np.linspace(0,1,10000)    #Anomalia excÃ©ntrica

M = (2*np.pi / T)*(t-to)     #Anomalia media

def fun(E):
    return (E) - (e*(np.sin(E))) - (M)





#---------------------------------------------------------------------------------------------------------------------------

""" ----------------------------------------- ConfiguraciÃ³n del grafico ----------------------------------------- """

#---------------------------------------------------------------------------------------------------------------------------

plt.style.use("classic")

plt.title("Orbitas ElÃ­pticas")

plt.xlabel("Anomalia ExcÃ©ntrica (E)")
plt.ylabel("PosiciÃ³n angular (Rad)")

plt.axhline( 0 , linestyle = "--" , color = "black")

plt.grid()

plt.plot( E , fun(E) , color="#EB211A")

plt.show()

#---------------------------------------------------------------------------------------------------------------------------

""" ----------------------------------------- Anomalia excentrica de el 1 de abril ----------------------------------------- """

#---------------------------------------------------------------------------------------------------------------------------

a = sp.optimize.bisect(fun,0.2,0.4)
print("La anomalia excentrica en abril 1 del 1968 es de",a)

#---------------------------------------------------------------------------------------------------------------------------

""" ----------------------------------------- Varaibles y paremetros usados (Periodo completo) ----------------------------------------- """

#---------------------------------------------------------------------------------------------------------------------------

E = np.linspace(0,1,1000) 

ExcentricidadesSec = []
ExcentricidadesNew = []
t = np.linspace(0,T,1000) 

for i in t:
    M = (2*np.pi / T)*(i)
    ExcentricidadesSec.append( sp.optimize.newton(fun , x0=i , x1=i+(T/100) ) )
    ExcentricidadesNew.append( sp.optimize.newton(fun , x0=i , x1=i+(T/100) ) )

#---------------------------------------------------------------------------------------------------------------------------

""" ----------------------------------------- ConfiguraciÃ³n del grafico (Periodo completo)----------------------------------------- """

#---------------------------------------------------------------------------------------------------------------------------

plt.style.use("classic")

plt.title("Orbitas ElÃ­pticas (Periodo completo)")

plt.xlabel("Tiempo (t)")
plt.ylabel("Anomalia excentrica (E)")

plt.axhline( 0 , linestyle = "--" , color = "black")

plt.grid()

plt.plot( t , ExcentricidadesSec , color="#28BE62")
plt.plot( t , ExcentricidadesNew , color="#D82487")

plt.show()

print("Podemos observar que el metodo de la secante y el metodo de newton se solapan por dar el mismo resultado")
```

```python
#-----Problema 2-----

#---------------------------------------------------------------------------------------------------------------------------

""" ----------------------------------------- Librerias ----------------------------------------- """

#---------------------------------------------------------------------------------------------------------------------------


import numpy as np
import matplotlib.pyplot as plt
import scipy as sp




#---------------------------------------------------------------------------------------------------------------------------

""" ----------------------------------------- Parametros y variables usadas----------------------------------------- """

#---------------------------------------------------------------------------------------------------------------------------


T = 0.5  

m = np.linspace(0,1,1000)
def mag(m):
    return np.tanh(m/T) - m

def magpuntofijo(m):
    return np.tanh(m/T)


#---------------------------------------------------------------------------------------------------------------------------

""" ----------------------------------------- Grafica de apoyo ----------------------------------------- """

#---------------------------------------------------------------------------------------------------------------------------

plt.style.use("classic")

plt.title("MagnetizaciÃ³n en funciÃ³n de la temperatura")

plt.ylabel("Temperatura (T)")
plt.xlabel("MagnetizaciÃ³n (m)")

plt.axhline( 0 , linestyle = "--" , color = "black")

plt.grid()

plt.plot( m , mag(m) , color="#28BE62")

plt.show()


#---------------------------------------------------------------------------------------------------------------------------

""" ----------------------------------------- ObtenciÃ³n de la raiz ----------------------------------------- """

#---------------------------------------------------------------------------------------------------------------------------

Raiz2S = sp.optimize.newton(mag , x0=0.8 , x1=1.0)
Raiz2N = sp.optimize.newton(mag , x0=0.9)
Raiz2P = sp.optimize.fixed_point(magpuntofijo , 0.8)

print("La magnetizaciÃ³n reducida por medio del metodo de Newton es", round(Raiz2N,4))
print("La magnetizaciÃ³n reducida por medio del metodo de Secante es", round(Raiz2S,4))
print("La magnetizaciÃ³n reducida por medio del metodo de Punto Fijo es", round(float(Raiz2P),4))





#---------------------------------------------------------------------------------------------------------------------------

""" ----------------------------------------- VariaciÃ³n de temperatura ----------------------------------------- """

#---------------------------------------------------------------------------------------------------------------------------


Temperatura = np.linspace(0.1,0.99,100)
Magred = []

for i in Temperatura:
    T = i
    Magred.append(sp.optimize.newton(mag , x0=0.9))




#---------------------------------------------------------------------------------------------------------------------------

""" ----------------------------------------- VariaciÃ³n de temperatura (Grafica)----------------------------------------- """

#---------------------------------------------------------------------------------------------------------------------------


plt.style.use("classic")

plt.title("MagnetizaciÃ³n Reducida")

plt.ylabel("Temperatura (T)")
plt.xlabel("MagnetizaciÃ³n Reducida(m)")

plt.xlim(0.05,1)
plt.ylim(0,1.05)
plt.axhline( 0 , linestyle = "--" , color = "black")

plt.grid()

plt.plot( Temperatura , Magred , color="#96BE28")

plt.show()
```

```python
#-----Problema 3-----

#---------------------------------------------------------------------------------------------------------------------------

"""Preguntas a contestar"""

print("---PREGUNTAS---")
print(" Si la energia libre de Gibbs es iguala a -10^5J\n 1.) Â¿CuÃ¡nto tendria que valer su temperatura?\n")  

#---------------------------------------------------------------------------------------------------------------------------

"""Librerias"""

import numpy as np
import matplotlib.pyplot as plt
import scipy as sp

#---------------------------------------------------------------------------------------------------------------------------

"""Parametros y variables usadas"""

R = 8.314                       #Constante de los gases
To = 4.444                      #Constante de referencia
G = -1e5                        #Energia libre de Gibbs

T = np.linspace(0.0001,1000,10000)          #Intervalo de temperatura

def fun3(T):                                #Funcion de la energia libre de Gibbs igualada a 0
    return ( -R*T * np.log( (T/To)**(5/2) ) ) - (G)


#---------------------------------------------------------------------------------------------------------------------------

"""    Obtencion de las raiz    """

Raiz3 = sp.optimize.bisect(fun3 , 890 , 910 , rtol = 1e-15 , maxiter = 10000)          #Raiz encontrada con el metodo de bisecciÃ³n

print("---CONCLUSIÃN---")
print("Por el metodo de BisecciÃ³n podemos confirmar que la energia libre de Gibbs iguala\n" \
"a -10^5 J cuando su temperatura vale aproximadamente",Raiz3,"Kelvin\n")


#---------------------------------------------------------------------------------------------------------------------------

"""    ConfiguraciÃ³n de el grafico    """

print("----GRAFICO----")
print("Para una mejor visualizaciÃ³n de la raiz encontrada")

plt.style.use("classic")

plt.title("Energia libre de Gibbs para un mol de hidrogeno")

plt.xlabel("Temperatura (K)")
plt.ylabel("Energia libre (J)")

plt.axhline( 0 , linestyle = "--" , color = "black")
plt.axvline( Raiz3 , linestyle = "--" , color = "black")

plt.grid()

plt.plot( T , fun3(T) , color="#EB211A")

plt.plot( Raiz3 , 0 , color="blue" , marker="o")

plt.show()
```

```python
#-----Problema 4-----

#---------------------------------------------------------------------------------------------------------------------------

""" ----------------------------------------- Preguntas a contestar ----------------------------------------- """

#---------------------------------------------------------------------------------------------------------------------------

print("---PREGUNTAS---")
print(" Si Io = 1")
print(" 1.) Â¿CÃ³mo son las graficas de I y su derivada?\n")
print(" 2.) Â¿Cuanto vale x cuando I(x) = Io/2? Usando el metodo de la secante y un xo = pi\n")
print(" 3.) Â¿Cuales son los maximos locales de la funciÃ³n I usando su derivada?\n")
print(" 4.) Â¿Cuales son los maximos locales de la funciÃ³n I usando su derivada (Regula Falsi)? h=0.5\n\n\n\n")




#---------------------------------------------------------------------------------------------------------------------------

""" ----------------------------------------- Librerias ----------------------------------------- """

#---------------------------------------------------------------------------------------------------------------------------

import numpy as np
import matplotlib.pyplot as plt
import scipy as sp




#---------------------------------------------------------------------------------------------------------------------------

""" ----------------------------------------- Varaibles, parametros y funciones usadas ----------------------------------------- """

#---------------------------------------------------------------------------------------------------------------------------

Io = 1                                                           #DistribuciÃ³n de intensidad inicial

x = np.linspace(-10 , 10 , 10000)                                #Intervalo de temperatura

def Intensidad(x):                                               #Funcion de la distribuciÃ³n de intensidad
    return (np.sin(x) / x)**2

def diffIntensidad(x):                                           
    return (2/x)*(np.cos(x) - (np.sin(x)/x))*(np.sin(x)/x)




#---------------------------------------------------------------------------------------------------------------------------

""" ----------------------------------------- ConfiguraciÃ³n de el grafico ----------------------------------------- """

#---------------------------------------------------------------------------------------------------------------------------

print("----GRAFICO DE LA INTENSIDAD----")
print("Para ver como se comporta la intesidad con respecto a su variable implicita")

plt.style.use("classic")

plt.title("Intensidad")

plt.xlabel("Variable implicita (x)")
plt.ylabel("Intensidad")

plt.axhline( 0 , linestyle = "--" , color = "black")

plt.grid()

plt.plot( x , Intensidad(x) , color="#EB211A")

plt.show()




#---------------------------------------------------------------------------------------------------------------------------

""" ----------------------------------------- ConfiguraciÃ³n de el grafico de la derivada ----------------------------------------- """

#---------------------------------------------------------------------------------------------------------------------------

print("\n\n\n\n----GRAFICO DE LA DERIVADA DE LA INTENSIDAD----")
print("Para ver como se comporta la tasa de cambio de la intensidad con respecto a su variable implicita")

plt.style.use("classic")

plt.title("Derivada de intensidad")

plt.xlabel("Variable implicita (x)")
plt.ylabel("Tasa de cambio de la intensidad")

plt.axhline( 0 , linestyle = "--" , color = "black")

plt.grid()

plt.plot( x , diffIntensidad(x) , color="#8AEB1A")

plt.show()




#---------------------------------------------------------------------------------------------------------------------------

""" ----------------------------------------- Raiz cuando I(x) = Io/2 ----------------------------------------- """

#---------------------------------------------------------------------------------------------------------------------------

def Int2(x):
    return Intensidad(x) - 1/2

Raiz4 = sp.optimize.newton( Int2 , x0=np.pi , x1=1)
print("\n\n\n\n----GRAFICO CUANDO I(x) = Io/2----")
print("Para que I(x) = Io/2 'x'(positivo) debe que ser igual a" , Raiz4)




#---------------------------------------------------------------------------------------------------------------------------

""" ----------------------------------------- ConfiguraciÃ³n de el grafico cuando I(x) = Io/2 ----------------------------------------- """

#---------------------------------------------------------------------------------------------------------------------------

plt.plot(x,Int2(x))

plt.style.use("classic")

plt.title("Intensidad menos Io/2")

plt.xlabel("Variable implicita (x)")
plt.ylabel("Intesidad")

plt.axhline( 0 , linestyle = "--" , color = "black")
plt.axvline( Raiz4 , linestyle = "--" , color = "black")

plt.plot( Raiz4 , 0 , color="#16E2D8" , marker="o")

plt.grid()

plt.show()




#---------------------------------------------------------------------------------------------------------------------------

""" ----------------------------------------- Max y Min con la derivada ----------------------------------------- """

#---------------------------------------------------------------------------------------------------------------------------

print("\n\n\n\n----MAXIMOS Y MINIMOS DE LA INTENSIDAD (SECANTE)---")

Raices = []
for i in range(0,20):
    if diffIntensidad(-9.9+i)*diffIntensidad(-8.9+i) <= 0:
        Raices.append(round(float(sp.optimize.newton(diffIntensidad,x0=-9.9+i , x1=-8.9+i)),5))
    else:
        pass

Raices = list(set(Raices))
Raices.sort()
Max = []
Min = []

for i in range(0,len(Raices)):
    if i%2 == 0:
        Min.append(Raices[i])
    else:
        Max.append(Raices[i])

print("\nLos maximos locales de nuestra funciÃ³n intensidad son:\n",Max)
print("\nLos minimos locales de nuestra funciÃ³n intensidad son:\n",Min)




#---------------------------------------------------------------------------------------------------------------------------

""" ----------------------------------------- MÃ©todo de regula falsi ----------------------------------------- """

#---------------------------------------------------------------------------------------------------------------------------

def regula(f , x0 , x1 , tol = 1e-15, maxiter = 10000):
    if f(x0)*f(x1)>0:
        print("El metodo no funciona")
    elif f(x0) == 0:
        return x0
    elif f(x1) == 0:
        return x1
    for i in range(0,maxiter):
        c = x1 - ( (f(x1)*(x1-x0)) / (f(x1)-f(x0)) )
        if abs(f(c)) < tol:
            return c
        elif f(x0) * f(c) < 0:
            x1 = c
        else:
            x0 = c




#---------------------------------------------------------------------------------------------------------------------------

""" ----------------------------------------- Uso de regula falsi ----------------------------------------- """

#---------------------------------------------------------------------------------------------------------------------------

print("\n\n\n\n----MAXIMOS Y MINIMOS DE LA INTENSIDAD (REGULA FALSI) h=0.5---")

Raices = []
for i in np.arange(0,20,0.5):
    if diffIntensidad(-9.9+i)*diffIntensidad(-8.9+i) <= 0:
        Raices.append(round(float(regula(diffIntensidad,x0=-9.9+i , x1=-8.9+i)),5))
    else:
        pass

Raices = list(set(Raices))
Raices.sort()

Max = []
Min = []

for i in range(0,len(Raices)):
    if i%2 == 0:
        Min.append(Raices[i])
    else:
        Max.append(Raices[i])

print("\nLos maximos locales (Regula Falsi) de nuestra funciÃ³n intensidad son:\n",Max)
print("\nLos minimos locales (Regula Falsi) de nuestra funciÃ³n intensidad son:\n",Min)


```

```python
#-----Problema 5-----

#---------------------------------------------------------------------------------------------------------------------------

""" ----------------------------------------- Librerias ----------------------------------------- """

#---------------------------------------------------------------------------------------------------------------------------


import numpy as np
import matplotlib.pyplot as plt
import scipy as sp




#---------------------------------------------------------------------------------------------------------------------------

""" ----------------------------------------- Variables, parametros y funciones usadas ----------------------------------------- """

#---------------------------------------------------------------------------------------------------------------------------

y = np.linspace(0.0000001,3.1e-6,10000)

h = 6.626e-34
kb = 1.381e-23
c = 2.998e8

def u(y,T):
    yo = ((h*c)/(kb*T))
    x = y / yo
    K = (8*np.pi*kb*T) / (yo**4)
    return K  *  ( 1 / ( (x**5) * ( (np.exp(1/x)) - 1) ) )

def dudy(y,T):
    yo = ((h*c)/(kb*T))
    x = y / yo
    K = (8*np.pi*kb*T) / (yo**4)
    return K*yo*( ( (5*x) + ( 1 - (5*x) ) * ( np.exp(1/x) ))  / ( (x**7) * ( ( (np.exp(1/x)) - 1 )**2 ) )  )

#---------------------------------------------------------------------------------------------------------------------------

""" ----------------------------------------- Grafica de la energia----------------------------------------- """

#---------------------------------------------------------------------------------------------------------------------------

plt.style.use("classic")

plt.title("Energia de la radiaciÃ³n electromagnetica")

plt.xlabel("Longitud de onda (m)")
plt.ylabel("Energia (J)")

plt.axhline( 0 , linestyle = "--" , color = "black")

plt.grid()

plt.ticklabel_format(style='sci', axis='both', scilimits=(0, 0))

plt.plot( y , u(y,4000) , color="#EB211A" , label="4000K")
plt.plot( y , u(y,5000) , color="#4DCE2D" , label="5000K")
plt.plot( y , u(y,6000) , color="#1A52EB" , label="6000K")
plt.legend()
plt.show()

#---------------------------------------------------------------------------------------------------------------------------

""" ----------------------------------------- Grafica de la derivada----------------------------------------- """

#---------------------------------------------------------------------------------------------------------------------------

plt.style.use("classic")

plt.title("Tasa de cambio de la energia de la radiaciÃ³n electromagnetica")

plt.xlabel("Longitud de onda (m)")
plt.ylabel("Energia (J)")

plt.axhline( 0 , linestyle = "--" , color = "black")

plt.grid()

plt.ticklabel_format(style='sci', axis='both', scilimits=(0, 0))

plt.plot( y , dudy(y,4000) , color="#EB211A" , label="4000K")
plt.plot( y , dudy(y,5000) , color="#4DCE2D" , label="5000K")
plt.plot( y , dudy(y,6000) , color="#1A52EB" , label="6000K")
plt.legend()
plt.show()




#---------------------------------------------------------------------------------------------------------------------------

""" ----------------------------------------- Derivada igualada a 0 con T = 5778K----------------------------------------- """

#---------------------------------------------------------------------------------------------------------------------------

def dudy0(y,T = 5778):
    yo = ((h*c)/(kb*T))
    x = y / yo
    K = (8*np.pi*kb*T) / (yo**4)
    return K*yo*( ( (5*x) + ( 1 - (5*x) ) * ( np.exp(1/x) ))  / ( (x**7) * ( ( (np.exp(1/x)) - 1 )**2 ) )  )

Raiz5N = sp.optimize.newton(dudy0 , x0=0.4e-6 , x1=0.6e-6)
Raiz5B = sp.optimize.bisect(dudy0 , 0.4e-6 , 0.6e-6)

print("Por medio de el metodo de newton (secante) obtenemos que su raiz es",Raiz5N,"\n" \
"Por el normal es mas atraida por la raiz de el cero que por la mas cercana.")
print("Por medio de el metodo de biseccion obtenemos que su raiz es",Raiz5B)


plt.ticklabel_format(style='sci', axis='both', scilimits=(0, 0))
plt.axhline( 0 , linestyle = "--" , color = "black")

plt.plot(y,dudy0(y))

plt.show()


#---------------------------------------------------------------------------------------------------------------------------

""" ----------------------------------------- Ley de el desplazamiento de Wien----------------------------------------- """

#---------------------------------------------------------------------------------------------------------------------------

def dudy4000 (y):
    return dudy(y,4000)
def dudy5000 (y):
    return dudy(y,5000)
def dudy6000 (y):
    return dudy(y,6000)

Max4000 = sp.optimize.bisect( dudy4000 , 0.1e-6 , 1e-6)
Max5000 = sp.optimize.bisect( dudy5000 , 0.1e-6 , 1e-6)
Max5778 = sp.optimize.bisect( dudy0 , 0.1e-6 , 1e-6)
Max6000 = sp.optimize.bisect( dudy6000 , 0.1e-6 , 1e-6)

print("Podemos comprobar que:","\n" \
"ymax4000 * 4000 = ",Max4000*4000,"\n" \
"ymax5000 * 5000 = ",Max5000*5000,"\n" \
"ymax5778 * 5778 = ",Max5778*5778,"\n" \
"ymax6000 * 6000 = ",Max6000*6000,"\n" \
"Que es lo que nos dice la ley de el\n" \
"desplazamiento de Wien.")
```

```python
#-----Problema 6-----

#---------------------------------------------------------------------------------------------------------------------------
""" ----------------------------------------- Variables, parametros y funciones usadas ----------------------------------------- """
#---------------------------------------------------------------------------------------------------------------------------

E = np.linspace(0.000001,19.999999,10000)*(1.60218e-19)
m = 9.11e-31
V = 20*(1.60218e-19)
w = 1e-9
h = 1.054e-34

def y1(E):
    return np.tan( (  ( (w**2)*(m)*(E) )/( 2*(h**2) )  )**(1/2) )
def y2(E):
    return ((V-E)/E)**(1/2)
def y3(E):
    return -(E/(V-E))**(1/2)


#----------------------------------------------------------------------------------------------------
""" -----------------------------------------GRAFICACIÃN----------------------------------------- """
#----------------------------------------------------------------------------------------------------
plt.style.use("classic")

plt.title("---")

plt.xlabel("---")
plt.ylabel("---")

plt.axhline( 0 , linestyle = "--" , color = "black")

plt.grid()

plt.ticklabel_format(style='sci', axis='both', scilimits=(0, 0))

plt.xlim(0,3e-18)
plt.ylim(-20,20)
plt.plot(E , y1(E),label="1")
plt.plot(E,y2(E),label="2")
plt.plot(E,y3(E),label="3")


plt.legend()
plt.show()


#----------------------------------------------------------------------------------------------------
""" -----------------------------------------VISUALIZACIÃN----------------------------------------- """
#----------------------------------------------------------------------------------------------------

print("Ignoraremos las asintotas")
print("Se cruzan aproximadamente en 0.2e-18 , 0.4e-18 , 0.7e-18 , 1.3e-18 , 1.8e-18 , 2.4e-18\n pero vamso a comprobarlo numericamente")

#----------------------------------------------------------------------------------------------------
""" --------------------------CALCULOS Y GRAFICA DE APOYO----------------------------------------- """
#----------------------------------------------------------------------------------------------------
def y4(E):
    return np.tan( (  ( (w**2)*(m)*(E) )/( 2*(h**2) )  )**(1/2) ) - ((V-E)/E)**(1/2)
primer = sp.optimize.bisect(y4 , 0.3e-18 , 0.5e-18)
segundo = sp.optimize.bisect(y4 , 1.2e-18 , 1.4e-18)
tercer = round(float(sp.optimize.bisect(y4 , 2.3e-18 , 2.5e-18)),20)

def y5(E):
    return np.tan( (  ( (w**2)*(m)*(E) )/( 2*(h**2) )  )**(1/2) ) + (E/(V-E))**(1/2)
cuarto = sp.optimize.bisect(y5,0.1e-18,0.4e-18)
quinto = sp.optimize.bisect(y5,0.6e-18,1e-18)
sexto = sp.optimize.bisect(y5,1.6e-18,2e-18)

plt.plot(E,y4(E))
plt.plot(E,y5(E))
plt.ylim(-10,10)
plt.axhline( 0 , linestyle = "--" , color = "black")

#----------------------------------------------------------------------------------------------------
""" -----------------------------------------CONCLUSIONES----------------------------------------- """
#----------------------------------------------------------------------------------------------------

print("Las intersecciones entre y1 y y2 son de:",primer,segundo,tercer)
print("Las intersecciones entre y1 y y3 son de:",cuarto,quinto,sexto)

print("Lo que concuerda con suerte con nuestras pobres mediciones a ojo")
```
