## Problema 1: Velocidad de motociclista
Una persona conduce una moto con rapidez constante $v_m$, mientras que un policia lo hace con aceleración constante $a_p$ (iniciando desde el reposo). Suponga que ambos movimientos inician en $t=0$, haga una gráfica para el movimiento de la moto y del policia, donde el usuario ingrese la velocidad (moto) y aceleración (policia). Además determine el instante $t$ en el cual se encuentran ambos. Ambos movimientos iniciarán en $x_i = 0$

**Pruebe**: para $v_m$ = 15m/s y $a_p$ = 3 m/s², el tiempo deberia ser 10 s


$$x_f = x_i + v_m\cdot t$$

$$x_f = x_i + v_o\cdot t + \frac{1}{2}a_p\cdot t^2$$



`````python
#Motociclista v.s Policia

"""
---------------------------------------------------------------------------------------------------
--------------------------------------LIBRERIAS USADAS---------------------------------------------
---------------------------------------------------------------------------------------------------
"""


import matplotlib.pyplot as plt
import numpy as np


"""
---------------------------------------------------------------------------------------------------
--------------------------------INSERCIÓN DE LOS DATOS---------------------------------------------
---------------------------------------------------------------------------------------------------
"""


Velocidad = int(input("Ingrese la velocidad del motociclista:"))
Aceleracion = int(input("Ingrese la aceleración del policia:"))


"""
---------------------------------------------------------------------------------------------------
--------------------------------------CALCULO DEL TIEMPO-------------------------------------------
---------------------------------------------------------------------------------------------------
"""


Tiempo = ( 2*Velocidad / Aceleracion)                                #Formula hallada al igualar sus posiciones
print("El policia alcanza al motociclista a los",Tiempo,"segundos.")


"""
---------------------------------------------------------------------------------------------------
--------------------------------------CREACION DE LOS EJES-----------------------------------------
---------------------------------------------------------------------------------------------------
"""


EjeT = np.linspace(0,2*Tiempo,100)              #Datos de el tiempo
EjeV = EjeT*Velocidad                           #Imagenes de la posición del motociclista
EjeA = (1/2)*(EjeT**2)*Aceleracion              #Imagenes de la posición de la policia


"""
---------------------------------------------------------------------------------------------------
--------------------------------------GRAFICACIÓN--------------------------------------------------
---------------------------------------------------------------------------------------------------
"""


plt.plot(EjeT,EjeV,"r--",label="Motociclista")          #Graficamos la posición del motociclista
plt.plot(EjeT,EjeA,"b--",label = "Policia")             #Graficamos la posición de la policia

plt.style.use("classic")                                #Estilo classic porque es bonito
plt.title("Motociclista v.s Policia")                   #Titulo, titulo de los ejes y leyendas
plt.xlabel("Tiempo (s)") 
plt.ylabel("Posición (m)") 
plt.legend()
plt.grid()                                              #Cuadricula

plt.show()

`````

## Problema 2: Potencial de Lennard-Jones
---

##2.
El potencial Lennard-Jones es un modelo simple que aproxima la interacción entre un par de moléculas neutras. Este está definido como:

$$
V(r) = 4\epsilon \left[ \left( \frac{\sigma}{r} \right)^{12} - \left( \frac{\sigma}{r} \right)^{6} \right]
$$

Donde:
- $V(r)$ es el potencial en función de la distancia interatómica $r$.
- $\epsilon$ es la profundidad del pozo de potencial, es decir, la energía mínima que describe la interacción entre dos partículas.
- $\sigma$ es la distancia a la cual el potencial es cero, también conocido como el parámetro de distancia.
- $r$ es la distancia entre las dos partículas.

a. Escriba una función para encontrar modelar el potencial LJ.


b. Usando $\sigma = \epsilon = 1$ demuestre que la distancia de equilibrio $r_m$ es aproximadamente $r_m = 2^{1/6}\sigma$. Para esto realice un gráfico del potencial, especificando los ejes. Trunque la serie en $r_c = 2.5\sigma$.


c. Verifique que $r_m = 2^{1/6}\sigma$ se cumple para diferentes valores de $\sigma$, dejando $\epsilon$ constante.




`````python
#Potencial de Lennard-Jones

"""
---------------------------------------------------------------------------------------------------
--------------------------------------LIBRERIAS USADAS---------------------------------------------
---------------------------------------------------------------------------------------------------
"""


import numpy as np
import matplotlib.pyplot as plt


"""
---------------------------------------------------------------------------------------------------
--------------------------------------ESTABLECER PARAMETROS----------------------------------------
---------------------------------------------------------------------------------------------------
"""


print("--------------------------------------------------")
print(" Vamos a graficar el potencial de Lennard - Jones\n Los parametros dados seran σ=1 y ϵ=1")
print("--------------------------------------------------")

Sigma = 1
E = 1

"""
---------------------------------------------------------------------------------------------------
--------------------------------------ESTABLECER LA FUNCIÓN----------------------------------------
---------------------------------------------------------------------------------------------------
"""

def Potencial(x):
    return 4*E*( (Sigma / x)**12 - (Sigma / x)**6 )                 #Función del potencial de Lennard-Jones

"""
---------------------------------------------------------------------------------------------------
--------------------------------------GRAFICO GENERAL----------------------------------------------
---------------------------------------------------------------------------------------------------
"""

Ejex = np.linspace(0.1,2.5*Sigma,100)                                                      #Lugar donde evaluaremos la función (PreImagen)
Ejey = Potencial(Ejex)                                                                     #Evaluando la función (Imagen)

plt.style.use("classic")                                                                   #Estilo clasico para estetica
plt.figure(figsize=(15,6))                                                                 #Tamaño del grafico

plt.xlim(0, 2.5*Sigma)                                                                     #Truncamiento de el grafico
plt.ylim(-abs(2*Potencial((2**(1/6)) * Sigma))),abs(2*Potencial((2**(1/6)) * Sigma))       #Truncamiento en el eje Y (Esto sabiendo que su minimo es [(2**(1/6)) * Sigma])


plt.title("Potencial de Lennard-Jones\nσ=1")                                               #Titulo, titulo de los ejes y cuadricula
plt.xlabel("Distancia intramolecular (nm)") 
plt.ylabel("Energia potencial (J)") 
plt.grid(True)

plt.plot(Ejex,Ejey,"g--",label="Variación de energia potencial")                                                            #Graficación
plt.plot((2**(1/6)) * Sigma,Potencial((2**(1/6)) * Sigma),"ro",label="Distancia de equilibrio")                             #Graficación de un punto para mejor visualización

plt.text((2**(1/6)) * Sigma * 1.1,Potencial((2**(1/6)) * Sigma) ,r"( $ 2^{1/6} , \mathbf{V}(2^{1/6})$ )", fontsize="14")    #Graficación de la coordenada explicita del punto
plt.legend(fontsize="12")
plt.show()

"""
---------------------------------------------------------------------------------------------------
--------------------------------------GRAFICO CON σ DISTINTO---------------------------------------
---------------------------------------------------------------------------------------------------
"""

print("----------------------------------------------------------------------------------")
print("Vamos a corroborar que sirve para diferentes valores de σ dejando ϵ constante en 1\nLas siguientes graficas corresponden a:\nσ=5      σ=30      σ=120      σ=930")
print("----------------------------------------------------------------------------------")

fig, axs = plt.subplots(2, 2, figsize=(13, 8))           #Se crea los subplots

#Grafica σ=5
Sigma = 5

Ejex = np.linspace(0.1,2.5*Sigma,100)
Ejey = Potencial(Ejex)

axs[0, 0].set_xlim(0, 2.5*Sigma)
axs[0, 0].set_ylim(-abs(2*Potencial((2**(1/6))*(Sigma))),abs(2*Potencial((2**(1/6))*(Sigma))))


axs[0, 0].set_title("Potencial de Lennard-Jones\nσ=5")
axs[0, 0].set_xlabel("Distancia intramolecular (nm)") 
axs[0, 0].set_ylabel("Energia potencial (J)") 
axs[0, 0].grid(True)

axs[0, 0].plot(Ejex,Ejey,"g--",label="Variación de energia potencial")
axs[0, 0].plot((2**(1/6)) * Sigma,Potencial((2**(1/6)) * Sigma),"ro",label="Distancia de equilibrio")

axs[0, 0].legend(fontsize="10")
axs[0, 0].text((2**(1/6)) * Sigma * 1.1,Potencial((2**(1/6)) * Sigma) ,r"( $ 2^{1/6}σ , \mathbf{V}(2^{1/6}σ)$ )", fontsize="10")


#Grafica σ=30


Sigma = 30

Ejex = np.linspace(0.1,2.5*Sigma,100)
Ejey = Potencial(Ejex)

axs[0, 1].set_xlim(0, 2.5*Sigma)
axs[0, 1].set_ylim(-abs(2*Potencial((2**(1/6))*(Sigma))),abs(2*Potencial((2**(1/6))*(Sigma))))


axs[0, 1].set_title("Potencial de Lennard-Jones\nσ=30")
axs[0, 1].set_xlabel("Distancia intramolecular (nm)") 
axs[0, 1].set_ylabel("Energia potencial (J)") 
axs[0, 1].grid(True)

axs[0, 1].plot(Ejex,Ejey,"g--",label="Variación de energia potencial")
axs[0, 1].plot((2**(1/6)) * Sigma,Potencial((2**(1/6)) * Sigma),"ro",label="Distancia de equilibrio")

axs[0, 1].legend(fontsize="10")
axs[0, 1].text((2**(1/6)) * Sigma * 1.1,Potencial((2**(1/6)) * Sigma) ,r"( $ 2^{1/6}σ , \mathbf{V}(2^{1/6}σ)$ )", fontsize="10")


#Grafica σ=120


Sigma = 120

Ejex = np.linspace(0.1,2.5*Sigma,100)
Ejey = Potencial(Ejex)

axs[1, 0].set_xlim(0, 2.5*Sigma)
axs[1, 0].set_ylim(-abs(2*Potencial((2**(1/6))*(Sigma))),abs(2*Potencial((2**(1/6))*(Sigma))))


axs[1, 0].set_title("Potencial de Lennard-Jones\nσ=120")
axs[1, 0].set_xlabel("Distancia intramolecular (nm)") 
axs[1, 0].set_ylabel("Energia potencial (J)") 
axs[1, 0].grid(True)

axs[1, 0].plot(Ejex,Ejey,"g--",label="Variación de energia potencial")
axs[1, 0].plot((2**(1/6)) * Sigma,Potencial((2**(1/6)) * Sigma),"ro",label="Distancia de equilibrio")

axs[1, 0].legend(fontsize="10")
axs[1, 0].text((2**(1/6)) * Sigma * 1.1,Potencial((2**(1/6)) * Sigma) ,r"( $ 2^{1/6}σ , \mathbf{V}(2^{1/6}σ)$ )", fontsize="10")


#Grafica σ=930


Sigma = 930

Ejex = np.linspace(0.1,2.5*Sigma,100)
Ejey = Potencial(Ejex)

axs[1, 1].set_xlim(0, 2.5*Sigma)
axs[1, 1].set_ylim(-abs(2*Potencial((2**(1/6)) * Sigma)),abs(2*Potencial((2**(1/6)) * Sigma)))


axs[1, 1].set_title("Potencial de Lennard-Jones\nσ=930")
axs[1, 1].set_xlabel("Distancia intramolecular (nm)") 
axs[1, 1].set_ylabel("Energia potencial (J)") 
axs[1, 1].grid(True)

axs[1, 1].plot(Ejex,Ejey,"g--",label="Variación de energia potencial")
axs[1, 1].plot((2**(1/6)) * Sigma,Potencial((2**(1/6)) * Sigma),"ro",label="Distancia de equilibrio r=2^(1/6)*σ")

axs[1, 1].legend(fontsize="10")
axs[1, 1].text((2**(1/6)) * Sigma * 1.1,Potencial((2**(1/6)) * Sigma) ,r"( $ 2^{1/6}σ , \mathbf{V}(2^{1/6}σ)$ )", fontsize="10")
plt.tight_layout()
plt.show()

"""
--------------------------------------------------------------Explicacion--------------------------------------------------------------------
Se crearon 4 graficas con el parametro sigma cambiado, el codigo es igual al de el grafico general solo que el sigma va variando y esta hecho
en subplots (Queda largo porque no los comprimi en un for:)
"""

"""
---------------------------------------------------------------------------------------------------
--------------------------------------Conclusión---------------------------------------------------
---------------------------------------------------------------------------------------------------
"""


print("--------------------------------------------------------------------")
print("Podemos comprobar que siempre el potencia es minimo cuando r=2^{1/6}")
print("--------------------------------------------------------------------")
`````

b. Realice un gráfico de barras, donde se muestre una comparación de la estatura media para niños (cm) y niñas (cm) en ambas categorias. Esto para los primeros 10 paises con mayor y menor altura, además agregue a Colombia.

`````python
#----Tablas con pandas----

"""
---------------------------------------------------------------------------------------------------
--------------------------------------LIBRERIAS USADAS---------------------------------------------
---------------------------------------------------------------------------------------------------
"""

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

"""
---------------------------------------------------------------------------------------------------
-------------CODIGO DE EL PROFESOR PARA IMPORTAR LAS TABLAS DE LAS PAGINAS WEB---------------------
---------------------------------------------------------------------------------------------------
"""


url = "https://worldpopulationreview.com/country-rankings/average-height-by-country"
html_content = pd.read_html(url)
alturas = html_content[0]

url = "https://worldpopulationreview.com/country-rankings/average-net-worth-by-country"
html_content = pd.read_html(url)
income = html_content[0]


"""
---------------------------------------------------------------------------------------------------
----------------------------LIMPIEZA DE LAS TABLAS DE LAS PAGINAS----------------------------------
---------------------------------------------------------------------------------------------------
"""


income = income.drop(columns=["Unnamed: 0"])                                      #Elimina columna vacia innecesario para el analisis
alturas = alturas.drop(columns=["Unnamed: 0"])

income = income.dropna(axis=1)                                                    #Elimina las columnas con datos vacios
alturas = alturas.dropna(axis=1)

income.columns = ["País", "GDP por adulto 2022"]                                  #Cambiando el nombre de las columnas al español
alturas.columns = ["País", "Altura promedio, chicos, 19 años (cm)","Chicas"]

alturas = alturas.iloc[:-1]                                                       #Elimina la fila "TOTAL" de el dataframe "alturas" ya que no la necesitamos

print("---------------------\nTabla de las alturas\n---------------------\n" , alturas)
print("---------------------\nTabla del dinero\n---------------------\n" , income)


"""
---------------------------------------------------------------------------------------------------
-------------------------------DATOS PARA LOS GRAFICOS DE BARRAS-----------------------------------
---------------------------------------------------------------------------------------------------
"""

Colombia = alturas[alturas["País"] == "Colombia"]                                 #Guardamos por separado los datos de Colombia

AltMax_B = pd.concat( [ alturas.nlargest(10,"Altura promedio, chicos, 19 años (cm)") ,  Colombia ] , ignore_index=True)  
AltMax_B = AltMax_B.sort_values(by="Altura promedio, chicos, 19 años (cm)" , ascending=False).drop("Chicas", axis=1) 

AltMax_G = pd.concat( [ alturas.nlargest(10,"Chicas") , Colombia ] , ignore_index=True)
AltMax_G = AltMax_G.sort_values(by="Chicas" , ascending=False).drop("Altura promedio, chicos, 19 años (cm)", axis=1)

AltMin_B = pd.concat( [ alturas.nsmallest(10,"Altura promedio, chicos, 19 años (cm)") ,  Colombia ] , ignore_index=True)
AltMin_B = AltMin_B.sort_values(by="Altura promedio, chicos, 19 años (cm)" , ascending=False).drop("Chicas", axis=1)

AltMin_G = pd.concat( [ alturas.nsmallest(10,"Chicas") , Colombia ] , ignore_index=True)
AltMin_G = AltMin_G.sort_values(by="Chicas" , ascending=False).drop("Altura promedio, chicos, 19 años (cm)", axis=1)


"""
--------------------------------------------------------------Explicacion--------------------------------------------------------------------
Creamos 4 dataframes distintos dependiendo de el genero, cada unos con los 10 datos de estatura mas grandes (n.largest) o con la estatura mas
baja (n.smallest) agregando a Colombia y ordenandolos de mayor a menor con (.sort_values). Tambien eliminamos la columna del genero contrario
para trabajar los datos mas facil.
"""


"""
---------------------------------------------------------------------------------------------------
------------------------------GRAFICOS DE BARRAS DE LAS ALTURAS------------------------------------
---------------------------------------------------------------------------------------------------
"""


fig, axs = plt.subplots( 2 , 2 , figsize=(50,50))                                  #Creamos 4 subplots

axs = axs.flatten()                                                                #Nos facilita los indice de los subplots a (0,1,2,3)

AltMod = [AltMax_B , AltMin_B , AltMax_G , AltMin_G]                               #Creamos listas para hacer los graficos con un for:
Nombres = ["Paises con chicos mas altos", "Paises con chicos mas bajos", "Paises con chicas mas altas", "Paises con chicas mas bajas"]
Colores = ["#180a9e"  , "#12e0f9" , "#cc1169" , "#e889b6"]
m = -1

plt.style.use("classic")                                                           #Estilo clasico porque se ve mas bonito

for i in AltMod:                                                                   #Va recorriendo la lista de dataframes
    m += 1                                                                             #Indica el indice del subplot
    axs[m].grid(True, linestyle='--', linewidth=2, color="gray" , axis="y" , zorder=0) #Modificaciones de la cuadricula (zorder es para que quede atras)
    axs[m].tick_params(axis='both', labelsize=25)                                      #Tamaño de los numeros de los ejes
    axs[m].tick_params(axis='x', rotation=-60)                                         #Rotamos los nombres de los paises para que no se solapen
    axs[m].set_yticks(np.linspace(150 , 190, 21))                                      #Hace que el eje Y tenga mas divisiones
    axs[m].set_title(Nombres[m] , fontsize = 50)                                       #Elige el nombre del grafico con la lista "Nombres"
    axs[m].set_xlabel("Paises" , fontsize=30)                                          #Fija el nombre en el eje X (Paises)
    axs[m].set_ylabel("Altura (cm)" , fontsize=30)                                     #Fija el nombre en el eje Y (Alturas (cm))
    axs[m].bar(i.iloc[:,0] , i.iloc[:,1] , color = Colores[m] , zorder=3)              #Selecciona los datos de el dataframe seleccionado
    axs[m].set_ylim(min(list(i.iloc[:,1])) -  4 , max(list(i.iloc[:,1])) + 4)          #Fija el tamaño en el eje Y para una buena lectura de datos

plt.show(block=True)


"""
---------------------------------------------------------------------------------------------------
---------------------------------RESULTADOS DE LOS GRAFICOS ---------------------------------------
---------------------------------------------------------------------------------------------------
"""


print("----------------------------------------------------------------------------------\nObservamos los datos de los paises con mayor y menor altura dependiendo del genero\n----------------------------------------------------------------------------------")


"""
---------------------------------------------------------------------------------------------------
-----------------------MERGE ENTRE LOS DATOS DEL DINERO Y AL ESTATURA------------------------------
---------------------------------------------------------------------------------------------------
"""


AltMax_B = pd.merge(AltMax_B, income, on="País", how="inner")                         
AltMax_G = pd.merge(AltMax_G, income, on="País", how="inner")
AltMin_B = pd.merge(AltMin_B, income, on="País", how="inner")
AltMin_G = pd.merge(AltMin_G, income, on="País", how="inner")


"""
--------------------------------------------------------------Explicacion--------------------------------------------------------------------
Modificamos los 4 dataframes anteriores para no almacenar mas datos uniendo las alturas con el dinero respectivo de cada país con el (.merge)
"""


"""
---------------------------------------------------------------------------------------------------
----------------------------GRAFICOS DE COMPARACION ESTARURA-DINERO--------------------------------
---------------------------------------------------------------------------------------------------
"""


fig, axs = plt.subplots( 2 , 2 , figsize=(50,50))                                       #Creamos 4 subplots
axs = axs.flatten()                                                                     #Nos facilita los indice de los subplots a (0,1,2,3)

AltMod = [AltMax_B , AltMin_B , AltMax_G , AltMin_G]                                    #Creamos nuevamente la lista para guardar cambios y hacer un for:
m=-1

for i in AltMod:                                                                        #Va recorriendo la lista de dataframes
    m += 1                                                                                  #Indica el indice del subplot
    Dinero=[]                                                                               #Esta lista y el for: siguiente se usa por el signo "$"
    for n in i["GDP por adulto 2022"]:                                                      #Al tenerlo es un "str" y necesitamos un "int" o "float"
        Dinero.append(float(n.replace(",","").replace("$","")))                                 #Elimina el simbolo de "$"
    axs[m].tick_params(axis='both', labelsize=25)                                           #Tamaño de los numeros de los ejes
    axs[m].set_title(Nombres[m] , fontsize = 50)                                            #Elige el nombre del grafico con la mimsa lista "Nombres"
    axs[m].set_xlabel("Altura (cm)" , fontsize=30)                                          #Fija el nombre en el eje X (Altura (cm))
    axs[m].set_ylabel("Dinero (USD)" , fontsize=30)                                         #Fija el nombre en el eje Y (Dinero (USD))
    axs[m].grid(True, linestyle='--', linewidth=2, color="black" , zorder=0)                #Modificaciones de la cuadricula (zorder es para que quede atras)
    axs[m].plot(i.iloc[:,1] , Dinero , "o" , color = Colores[m] , markersize=20 , zorder=3) #Selecciona los datos de el dataframe seleccionado

plt.show(block=True)


"""
---------------------------------------------------------------------------------------------------
---------------------------------RESULTADOS DE LA RELACIÓN ----------------------------------------
---------------------------------------------------------------------------------------------------
"""

print("----------------------------------------------------------------------------------\nPodemos observar que su relación es casi nula, los puntos parecen estar incluso puestos al azar probando que no estan relacionados para nada\n----------------------------------------------------------------------------------")
`````
