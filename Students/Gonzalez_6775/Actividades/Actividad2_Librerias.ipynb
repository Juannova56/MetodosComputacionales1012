## 1.
Una persona conduce una moto con rapidez constante $v_m$, mientras que un policia lo hace con aceleración constante $a_p$ (iniciando desde el reposo). Suponga que ambos movimientos inician en $t=0$, haga una gráfica para el movimiento de la moto y del policia, donde el usuario ingrese la velocidad (moto) y aceleración (policia). Además determine el instante $t$ en el cual se encuentran ambos. Ambos movimientos iniciarán en $x_i = 0$

**Pruebe**: para $v_m$ = 15m/s y $a_p$ = 3 m/s², el tiempo deberia ser 10 s


$$x_f = x_i + v_m\cdot t$$

$$x_f = x_i + v_o\cdot t + \frac{1}{2}a_p\cdot t^2$$



`````python
import matplotlib.pyplot as plt
import numpy as np

v=float(input('ingresa la velocidad de la moto: '))
a=float(input('ingresa la aceleracion de la policia: '))

encuentro=2*v/a
t=np.linspace(0,encuentro+2,100)
moto=v*t
policia=(a*t**2)/2
x_encuentro = v * encuentro


plt.plot(t, moto, label="Moto (velocidad constante)", color='blue')
plt.plot(t, policia, label="Policía (aceleración)", color='red')
plt.axvline(encuentro, color='green', linestyle='--', alpha=0.9, label=f'encuentro en t=,{encuentro} s')
plt.scatter(encuentro, x_encuentro, color='black', zorder=5, label='punto de encuentro')
plt.title("Posicion vs tiempo")
plt.xlabel("Tiempo (s)")
plt.ylabel("Posicion (m)")
plt.legend()
plt.grid(True)
plt.show()
`````

##2.
El potencial Lennard-Jones es un modelo simple que aproxima la interacción entre un par de moléculas neutras. Este está definido como:

$$
V(r) = 4\epsilon \left[ \left( \frac{\sigma}{r} \right)^{12} - \left( \frac{\sigma}{r} \right)^{6} \right]
$$

Donde:
- $V(r)$ es el potencial en función de la distancia interatómica $r$.
- $\epsilon$ es la profundidad del pozo de potencial, es decir, la energía mínima que describe la interacción entre dos partículas.
- $\sigma$ es la distancia a la cual el potencial es cero, también conocido como el parámetro de distancia.
- $r$ es la distancia entre las dos partículas.

a. Escriba una función para encontrar modelar el potencial LJ.


b. Usando $\sigma = \epsilon = 1$ demuestre que la distancia de equilibrio $r_m$ es aproximadamente $r_m = 2^{1/6}\sigma$. Para esto realice un gráfico del potencial, especificando los ejes. Trunque la serie en $r_c = 2.5\sigma$.


c. Verifique que $r_m = 2^{1/6}\sigma$ se cumple para diferentes valores de $\sigma$, dejando $\epsilon$ constante.




`````python
import matplotlib.pyplot as plt
import numpy as np

def potencial (r,e,o):
  return 4*e*(((o/r)**12)-((o/r)**6))

e=1
o=1
rc=2.5
rm=2**(1/6)*o
vrm=potencial(rm,e,o)
r=np.linspace(0.9,rc,500)
v=potencial(r,e,o)
plt.plot(r,v, color='blue', label=f"e= {e}, o= {o}")
plt.axvline(rm,linestyle='--', label=f'distancia de equilibrio {rm}', color='red')
plt.scatter(rm, vrm, color='black', alpha=0.9,)
plt.title("potencial lennard jones")
plt.xlabel('distancia entre las dos particulas (r)')
plt.ylabel('potencial v(r)')
plt.legend()
plt.grid()
plt.show()

for o in range (2,6):
  rm=2**(1/6)*o
  r=np.linspace(0.8,rm+2,500)
  vrm=potencial(rm,e,o)
  v=potencial(r,e,o)
  plt.plot(r,v, color='blue', label=f"e= {e}, o= {o}")
  plt.axvline(rm,linestyle='--', label=f'distancia de equilibrio rm= {rm}', color='red')
  plt.scatter(rm, vrm, color='black', alpha=0.9,)
  plt.title("potencial lennard jones")
  plt.xlabel('distancia entre las dos particulas (r)')
  plt.ylabel('potencial v(r)')
  plt.xlim(rm - 1.5, rm + 1.5)
  plt.ylim(vrm - 1, 2)
  plt.legend()
  plt.grid()
  plt.show()

`````

b. Realice un gráfico de barras, donde se muestre una comparación de la estatura media para niños (cm) y niñas (cm) en ambas categorias. Esto para los primeros 10 paises con mayor y menor altura, además agregue a Colombia.

Ahora, empleando los datos de altura para los 10 mayores (19 años) y 10 menores (19 años). Realice un gráfico donde relacione la altura media con el GDP y concluya si existe alguna relación.


`````python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

url = "https://worldpopulationreview.com/country-rankings/average-height-by-country"
html_content = pd.read_html(url)
alturas = html_content[0]

alturas = alturas.drop(columns=[col for col in alturas.columns if "in" in col.lower()])
alturas.columns = ["Ranking", "País", "Altura promedio hombres (cm)", "Altura promedio mujeres (cm)"]

print(alturas)
top10 = alturas.nlargest(10, "Altura promedio hombres (cm)")

bottom10 = alturas.nsmallest(10, "Altura promedio hombres (cm)")

colombia = alturas[alturas["País"] == "Colombia"]
otros_paises = pd.concat([top10, bottom10])

if "Colombia" not in otros_paises["País"].values:
    otros_paises = pd.concat([otros_paises, colombia])

otros_paises = otros_paises.reset_index(drop=True)

x = np.arange(len(otros_paises["País"]))
width = 0.35

fig, ax = plt.subplots(figsize=(14, 6))
bars1 = ax.bar(x - width/2, otros_paises["Altura promedio hombres (cm)"], width, label="Hombres")
bars2 = ax.bar(x + width/2, otros_paises["Altura promedio mujeres (cm)"], width, label="Mujeres")

ax.set_ylabel("Altura promedio (cm)")
ax.set_title("Comparación de estatura promedio por país (Hombres vs Mujeres)")
ax.set_xticks(x)
ax.set_xticklabels(otros_paises["País"], rotation=45, ha='right')
ax.legend()
plt.tight_layout()
plt.show()
`````
