El **Teorema de Laplace** es un método para calcular el determinante de una matriz cuadrada, particularmente útil para matrices de orden mayor a 2. Este teorema se basa en la expansión del determinante por los elementos de una fila o una columna cualquiera.



$$
\det(A) = \sum_{j=1}^n (-1)^{1+j} a_{1j} M_{1j}
$$

donde:
- $a_{1j}$ es el elemento de la primera fila y columna $j$.
- $M_{1j}$ es el menor asociado al elemento $a_{1j}$, es decir, el determinante de la submatriz de $3 \times 3$ que se obtiene al eliminar la fila 1 y la columna $j$.
- $(-1)^{1+j}$ es el signo correspondiente al cofactor del elemento $a_{1j}$.

Podemos realizar una función recursiva para el cálculo del determinante, sabiendo que el valor del determinante de una matriz de orden uno es el único elemento de esa matriz, y el de una matriz de orden superior a uno es la suma de cada uno de los elementos de una fila o columna por los Adjuntos a ese elemento, como en la función recursiva se emplea la misma función definida el cálculo lo haremos por Menor complementario, un ejemplo desarrollado por la primera fila sería:

$$
   \det (A_{j,j}) =
   \left \{
   \begin{array}{llcl}
      si & j = 1 & \to & a_{1,1} \\
                                 \\
      si & j > 1 & \to & \displaystyle \sum_{k=1}^j \; (-1)^{(1+k)} \cdot a_{1,k} \cdot \det( \alpha_{1,k})
   \end{array}
   \right .
$$

Realice una función que encuentre el determinante de una matriz usando la recursividad aqui planteada, explique explicitamente su código

```python
#Bien! nada mejor que una funcion recursiva, pues evita por si mismo ciclos internos!
def det(A):
  '''
  Calcula el determinante de una matriz a través de métodos recursivos.
  Entradas:
    A: Matriz CUADRADA jxj (ndarray)
  Salidas:
    det(A): Determinante de A (float)
  '''
  determinante=0 #Aqui almacenamos la respuesta.

  j=np.shape(A)[0] #Medimos las dimensiones de A (da igual si 0, o 1, porque es cuadrada)
  if j==1:         #Y ponemos nuestra condición inicial
    return A[0,0]


  else:               #sino, que realice la suma
    for k in range(0,j):
      subA=A #Construimos aca una especie de "copia", subA será nuestra α1,k
      subA=np.delete(subA,0,axis=0) #Hacemos el slizing (α1,k), donde siempre quita la primera fila
      subA=np.delete(subA,k,axis=1) #Y la columna K (que queda 0, porque en la iteracion anterior k-1 ya despaercio!)

      #Finalmente, solo hacemos la suma  (k+1, pues k empieza en 0)
      determinante+=((-1)**(1+(k+1)))*(A[0,k])*det(subA) #y VOILÀ! nuestra recursividad
      #Note que cuando entra det(subA), ahora es una matriz de orden j-1, y seguira el ciclo hasta ser de orden 1
      #proporcionando un valor para la iteración anterior, y anterior, y anteiror, volviendo a nuestro estado final!
  return determinante
```

#3 Método de Gauss - Seidel

Sea \$A\in\mathbb{R}^{n\times n}\$ no singular y sea \$b\in\mathbb{R}^n\$.
Descomponga \$A\$ como

$$
A \;=\; D \;+\; L \;+\; U,
$$

donde

* \$D\$ es la matriz diagonal de \$A\$,
* \$L\$ es la parte estrictamente triangular inferior,
* \$U\$ es la parte estrictamente triangular superior.

El algoritmo de Gauss - Seidel reorganiza el sistema \$Ax=b\$ como

$$
x \;=\; (D+L)^{-1}\bigl(b \;-\; Ux\bigr),
$$

y genera la sucesión

$$
x_i^{(k+1)}
= \frac{1}{a_{ii}}
\Bigl(b_i - \sum_{j<i} a_{ij}\,x_j^{(k+1)} - \sum_{j>i} a_{ij}\,x_j^{(k)}\Bigr),
\qquad i=1,\dots,n.
$$

Implemente una función `gauss_seidel(A, b, tol=1e-7, max_iter=100)` que:
   * Realice las iteraciones hasta que
     $\lVert x^{(k+1)}-x^{(k)}\rVert_\infty<\text{tol}$
     o se alcance `max_iter`;
   * devuelva el vector solución aproximado \$x\$, el número de iteraciones realizadas y la norma del último residuo.

Incluya una documentación clara.

Luego,

   * Genere una matriz aleatoria \$5\times5\$ (por ejemplo, con `np.random.rand`) y un vector \$b\$ aleatorio.
   * Resuelva \$Ax=b\$ con su función; calcule el error relativo frente a `numpy.linalg.solve`.
   * Estime igualmente el error respecto a la solución obtenida mediante \$x=A^{-1}b\$ (usando `numpy.linalg.inv`).
   * Presente las normas de los residuos y los errores relativos.

```python
#Bien! construyamos nuestro Gauss_Seidel
def gauss_seidel(A, b, tol=1e-7, max_iter=100):
  '''
  Resuelve el sistema Ax=b, a través del método Gauss_Seidel de manera iterativa para garantizar una convgerencia
  Entradas:
    A: Matriz de coeficientes (nxn)(ndarray)
    b: vector de terminos independientes (array)
    tol(opcional): tolerancia del calculo, estandar=1e-7 (float)
    maxiter(opcional): Iteraciones maximas antes de detenerse, estandar=100 (int)
  Salidas:
    xnew: vector de soluciones aproximado(array)
    iteraciones realizadas: cuantas iteraciones tomo el programa (float)
    norma del residuo: Ver que tanto se aleja el resultado a traves de r=b-Ax, donde ∥r∥ debe tender a 0 (float)
  El programa lo devuelve en forma de tupla como sigue:
  xnew,iteraciones,norma del residuo.

  DISCLAIMER: La convergencia del método esta garantizado para matrices de coeficientes diagonalmente dominantes
  y simetrica definida positivas. Si no es el caso, no hay garantía de la convergencia.


  '''
  x=np.zeros_like(b,dtype=float) #Esta es nuestra suposición inicial, un clasico y conveniente (0,0,...0)
  xnew=np.zeros_like(b,dtype=float) #Y nuestro x**k+1 (es decir, los terminos siguientes)
  #Luego, procedemos a hallar los nuevos x's, con los x's anteriores
  n=np.shape(A)[0]  #las dimensiones de la matriz nxn
  iteraciones=0

  while iteraciones < max_iter: #Nuestro tope

    for i in range(n):
      suma=0   #Cada que cambiemos de i, reiniciamos la suma

      for j in range(n):
        if j<i:
          suma+= (A[i][j])*xnew[j]       #En esta sección se hacen todas las sumas con i estático
        elif j>i:                        #Proporcionamos la condiciones
          suma+= (A[i][j])*x[j]

      xnew[i]=((b[i])-suma)/(((A[i][i]))) #Voila nuestro nuevo x[i]!


    error=np.linalg.norm( xnew - x ,ord=np.inf) #Calculamos el error
    residuo= (b-(A@xnew))                #Y el residuo
    if error<tol:
      return xnew,iteraciones,np.linalg.norm(residuo,ord=np.inf) #tal como se indica en las "Salidas"
    else:
      x[:]=xnew #si no logro convergencia, actualizamos y seguimos
      iteraciones+=1
  print(f'Máximas iteraciones alcanzadas')
  return xnew, iteraciones, np.linalg.norm(b - (A@xnew), ord=np.inf)
```

```python
''' Probemos nuestro nuevo auto deportivo! '''
#Importante! Recordemos que esta bella función, garantiza una convergencia para matrices diagonalmente dominantes, y...
#... ademas,simetrica definida positiva! Tal como el caso a continación:

A = np.array([
    [4, -1, 0, 0],
    [-1, 4, -1, 0],   #Una matriz que cumple los criterios...
    [0, -1, 4, -1],
    [0, 0, -1, 3] ], dtype=float)

b = np.array([15, 10, 10, 10], dtype=float)

#Notemos entonces, como con esta matriz, nuestro metodo es terriblemente eficiente!!

xReal=np.linalg.solve(A,b)
xAprox=gauss_seidel(A,b)[0]
errRelativo=((np.linalg.norm(xReal-xAprox))/(np.linalg.norm(xReal)))
print(f'Con Matrices diagonalmente dominantes, obtenemos un error relativo del {errRelativo*100}%')
```

```python
#Tambien podemos analizar la situación con la xReal asociada a la inversa!
xRealinversa=np.linalg.inv(A)@b
xAprox=gauss_seidel(A,b)[0]
errRelativoInversa=((np.linalg.norm(xRealinversa-xAprox))/(np.linalg.norm(xRealinversa)))
print(f'Con Matrices diagonalmente dominantes, obtenemos un error relativo del {errRelativoInversa*100}%')

#Podemos ver que ambos resultados son bastante cercanos, siendo el primer metodo, muy ligeramente mejor (muy ligeramente!)
```

```python
#Finalmente, organicemos nuestras conclusiones
datos={'Norma del Residuo':[gauss_seidel(A,b)[2]],'Iteraciones Necesarias':[gauss_seidel(A,b)[1]],
       'Error relativo (%) con  .solve':[errRelativo*100],'Error relativo(%) con inversa':[errRelativoInversa*100]}
tabla=pd.DataFrame(datos)
tabla
```

```python
''' Tal como sospechamos, con matrices diagonalmente dominantes, funciona de maravilla, en muy pocas iteraciones'''

#Ahora bien, con matrices... en general, nuestro programa, suele tener problemas de convergencia muy grandes! Observemos.
A=np.random.rand(5,5)
b=np.random.rand(5)      #Matrices en general

'Ahora, procedemos a hacer exactamente el mismo análisis'
xReal=np.linalg.solve(A,b)
xAprox=gauss_seidel(A,b)[0]
errRelativo=((np.linalg.norm(xReal-xAprox))/(np.linalg.norm(xReal)))
print(f'Con Matrices diagonalmente dominantes, obtenemos un error relativo del {errRelativo*100}%')
```

#4 Método de potencias para el valor propio dominante

Sea \$A\in\mathbb{R}^{n\times n}\$ diagonalizable con valor propio dominante \$\lambda\_{\max}\$ (en magnitud) y vector propio asociado \$v\_{\max}\$.

El método de potencias genera, a partir de un vector inicial \$q^{(0)}\neq 0\$, la sucesión

$$
q^{(k+1)} \;=\; \frac{A\,q^{(k)}}{\lVert A\,q^{(k)}\rVert_2},
\qquad
\lambda^{(k+1)} \;=\; (q^{(k+1)})^{\!\top} A\, q^{(k+1)},
$$

que converge a \$v\_{\max}/\lVert v\_{\max}\rVert\_2\$ y a \$\lambda\_{\max}\$ respectivamente, bajo hipótesis estándar.

Implemente `power_method(A, tol=1e-7, max_iter=1000)` que:

   * Acepte matrices reales cuadradas,
   * Devuelva \$\lambda\_{\max}\$, el vector propio normalizado \$v\_{\max}\$, el número de iteraciones y la última variación relativa de \$\lambda\$,
   * detenga la iteración cuando
     $\bigl|\lambda^{(k+1)}-\lambda^{(k)}\bigr|<\text{tol}\,|\lambda^{(k+1)}|$
     o se alcance `max_iter`.

Luego,
   * Genere una matriz simétrica aleatoria \$6\times6\$ (por ejemplo, \$A = (M+M^\top)/2\$ con \$M\$ aleatoria).
   * Aplique su `power_method` y compare \$\lambda\_{\max}\$ y \$v\_{\max}\$ con los resultados de `numpy.linalg.eig`.

```python
#Bien! El plan es sencillo, generamos un vector aleatorio, y nos vamos  a la convergencia!

def power_method(A,tol=1e-7,max_iter=1000):
  '''
  Calcula el valor propio dominante de una matriz cuadrada a través del método de potencias, al igual
  que su vector propio dominante normalizado.
  Entradas:
    A: matriz cuadrada (ndarray nxn)
    tol(opcional): Tolerancia establecida (metodo iterativo) (estandar=1e-7) (float)
    max_iter(opcional): màximas iteraciones a realizar (estandar=1000) (int)
  Salidas:
    valor propio dominante (float), vector propio dominante normalizado (array), numero de iteraciones (float),
    última variación relativa del valor propio (float (%)).
  '''
  #Primero, creamos nuestro vector aleatorio
  n=len(A)#Como es una matriz cuadrada, podemos aprovechar el len
  q=np.zeros(n) #Generamos nuestro vector muestra  y...
  for k in range(n):
    q[k]=np.random.random()  #listo! tenemos un vector muestra TOTALMENTE ALEATOREO

  qnew=np.copy(q) #y hacemos nuestra copia para la iteración futura.
  lambdy=0           #y nuestros valores de lambda anterior y nuevo!
  lambdynew=0

  for i in range(max_iter): #y ahora si, comencemos a iterar.
    qnew=(A@q)/(np.linalg.norm((A@q), ord=2) ) #nuestro vector propio
    lambdynew=(np.transpose(qnew))@(A@qnew)    #y nuestro valor propio

    if np.linalg.norm(lambdynew-lambdy)<tol:
      return lambdynew, qnew, i, (((np.abs(lambdynew-lambdy))/(np.abs(lambdynew)))*100)
      #  valor propio, vector propio, iteraciones y variación relativa del valor obtenido
    else:
      q[:]=qnew   #En caso contrario, actualizamos (forma iterativa convencional)
      lambdy=lambdynew




```

```python
#Bien! probemos nuestro juguete nuevo!
#construyamos nuestra matriz simétrica aleatoria:
M=np.random.rand(6,6)
A=(M+np.transpose(M))/2
apr=power_method(A) #ejecutamos y..

lamAprx=apr[0]   #comenzamos a extraer!
lamReal=np.max(np.linalg.eigh(A)[0])

qAprx=apr[1]
qPropios=np.linalg.eigh(A)[1]
qReal=qPropios[:,-1]

#Hora de Comparar!!!
```

Verifique que cualquier matriz hermitiana de 2 × 2 $ L $ puede escribirse como una suma de cuatro términos:

$$ L = a\sigma_x + b\sigma_y + c\sigma_z + dI $$

donde $ a $, $ b $, $ c $ y $ d $ son números reales.

Las cuatro matrices de Pauli son:

$$ \sigma_x = \begin{pmatrix} 0 & 1 \\ 1 & 0 \end{pmatrix}, \quad \sigma_y = \begin{pmatrix} 0 & -i \\ i & 0 \end{pmatrix}, \quad \sigma_z = \begin{pmatrix} 1 & 0 \\ 0 & -1 \end{pmatrix}, \quad I = \begin{pmatrix} 1 & 0 \\ 0 & 1 \end{pmatrix} $$




```python
#Bien,comencemos por nuestra definición: ¿Que es una matriz hermitiana?
#Es una matriz H tal que H=H†. donde H†, es la transpuesta de la conjugada de H
#La conjugada de H, no es mas que cambiar el signo de los elementos complejos de la matriz
```

```python
#Para el proceso, empleemos una adaptación de nuestro molde del punto uno.

def Hermitiana(N):
  '''
  Demuestra que cualquier matriz hermitiana de 2×2 L  puede escribirse como una suma de cuatro términos
  Entradas:
    N: numero de iteraciones(int)
  Salidas:
    Conclusion: informa si la propiedad se cumple (str)
  '''

  σx=np.array([[0,1],                  #Creamos entonces nuestras matrices de Pauli
              [1,0]],dtype=complex)
  σy=np.array([[0,-1j],
              [1j,0]],dtype=complex)
  σz=np.array([[1,0],
               [0,-1]],dtype=complex)
  I=np.eye(2)

  o=0
  for i in range (N): #(N asociado al nùmero de iteraciones)
    a,b,c,d=np.random.rand(),np.random.rand(),np.random.rand(),np.random.rand() #Luego, obtenemos nuestros numeros reales aleatorios
    L=a*σx+b*σy+c*σz+d*I # Y voilà! Nuestra maravillosa matriz L, ahora confirmemos su "Hermiticidad"

    if np.isclose((L), (np.transpose(np.conjugate(L))) ).all()==True: #si es hermitica...
      o+=1  #Sumamos 1

  if o==N:  #Y aqui confirmamos!
    return print('cualquier matriz hermitiana de 2×2 L  puede escribirse como una suma de matrices de Pauli')
  else:
    return print('L como suma de matrices de Pauli no siempre genera matrices Hermiticas')
```

En el área de las ciencias exactas, los sistemas de ecuaciones, transformaciones, y análisis de datos se valen del algebra lineal para un eficiente desarrollo de sus planteamientos, sin embargo, en nuestro mundo moderno, la cantidad de datos que debemos procesar, o los numerosos cálculos que se deben realizar, hacen que las soluciones analíticas se vuelvan complicadas, e inclusive, en muchas ocasiones, en absoluto inviables, donde nuestros métodos computacionales se vuelven indispensables, tales como:

-**Operaciones generales matriciale**s: Para el correcto funcionamiento de todo el análisis lineal, se hace una breve mención al algebra matricial, con funciones asociadas al producto interno, multiplicación entre matrices, trazas, transpuestas, etc. (Asociado a Numpy)

-**np.linalg.solve**():  Esencial en problemas físicos generales, pues resuelve sin problema, sistemas de ecuaciones lineales, proporcionando soluciones inmediatas, como lo pueden ser circuitos eléctricos a través de las leyes de Kirchhoff, o mismo, campos mas avanzados como la cuántica permitiendo solucionar matrices complejas.

-**np.linalg.inv()**: Conocer la inversa de una matriz, permite acercamientos diferentes a solución de sistemas de ecuaciones de manera x= A**-1*b, al igual que análisis de transformaciones en caso de querer aprovechar su inyectividad para regresar al estado original previo a la transformación. A su vez, hallar la inversa de una matriz analíticamente conlleva vastos cálculos, siendo una ventaja la eficacia computacional.

-**np.linalg.pinv**(): No todas las matrices tienen inversa, sin embargo, nuestros métodos suelen tener acercamientos a través de aproximaciones realmente precisas, donde, gracias a esta función, podemos aprovechar los maravillosos usos de las matrices inversas, ¡mismo para matrices singulares!, desde sistema de ecuaciones, hasta regresiones lineales más complejas.

-**np.linalg.det**():  Es una pieza clave en la columna del algebra lineal, y calcularlo analíticamente, suele requerir numerosos pasos, sobre todo cuando las matrices crecen de dimensión. Su calculo computacional da basta información sobre el carácter del sistema, habla sobre posibles soluciones en sistemas de ecuaciones (única, infinitas, ninguna) y a su vez, es un método alternativo para calcular inversas.

-**np.linalg.eigh()**: Obtener los valores y vectores propios de una matriz, representan información clave en análisis de diferentes fenómenos; su carácter inherente se encuentra en la mecánica cuántica, en componentes como lo son la energía y la posición, al igual que se les suele asociar al estudio de la simetría en sistemas.

-**np.isclose().all**: No es directamente exclusivo del algebra lineal, pero es clave para comparar matrices, pues en muchas ocasiones, los cálculos aproximativos suelen ir un par de decimales hacia arriba, o hacia abajo, haciendo que dos matrices prácticamente idénticas no se vean reflejadas como tal al realizar A==B, por lo cual, este método es de sobremanera practico para un análisis general, sobre todo en casos aproximativos.

-**Descomposición LU**:  Permite descomponer una matriz A, en formas lower y upper, que, en diversos casos, el calculo Ax=b, puede llegar a ser costoso computacionalmente, por lo cual, aprovechando que las matrices LU son triangulares, los costos de calculo se reducen de sobre manera. también facilitan los cálculos de determinantes e inversas. (Note que, internamente .solve realiza esta clase de descomposiciones!, incluyendo extras como el pivoteo )

Se realiza una mención honorifica a los métodos iterativos para hallar solución a sistemas de ecuaciones, como lo son Jacobi y Gauss-Seidel , que son métodos muy eficientes en aproximaciones, pero son especializados, es decir, las matrices que analizan deben cumplir condiciones muy específicas, como lo pueden ser matrices diagonales dominantes, mientras que en casos de matrices aleatorias suelen no converger.
