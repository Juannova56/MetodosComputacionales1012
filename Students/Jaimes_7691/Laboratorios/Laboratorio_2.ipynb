### Primer problema

```python
def myint(x):
    if len(x) != 16:
        raise ValueError('El binario debe tener 16 bits')
    binario = list(x)
    entero=0
    for i in range(16):
        bit = int(binario[15-i])
        entero += bit * (2 **i)
    return entero

#Binario de 16 bits
a='0000001000011101'
print('El resultado con myint: ', myint(a),
    '\nEl resultado con int(): ', int(0b1000011101))
```

### Segundo problema

```python
def number64(x):

    signo = int(x[0])
    exponente = int(x[1:12], 2) - 1023
    mantisa = x[12:]
    mantisa_transformada = 1

    for i in range(len(mantisa)):
        if mantisa[i] == '1':
            mantisa_transformada += 2 ** (-(i + 1))
    return ((-1)**signo) * mantisa_transformada * (2**exponente)

b = "0100000000111011100100001111111111111111111111111111111111111111"
number64(b)
```

### Tercer problema

```python
#Aproximación básica

def arctan_basica(tolerancia, maxiter=100000):
    suma=0
    x=1
    n=0
    for i in range(1, maxiter+1):
        sm= ((-1)**(i+1)) * (x**(2*i-1)) / (2*i-1)
        suma += sm
        n += 1
        if abs(4*suma - np.pi) < tolerancia:
            return 4*suma, n
    return f'Se llego al maximo de iteraciones {maxiter} llego a {4*suma}'

print('<<<<<<<<<<Aproximación básica de pi>>>>>>>>>>\n')

piaprox, iteraciones = arctan_basica(1e-3)

print('El numero mínimo de términos necesarios para que |4Pn(1)−π| < 10^-3 es:\n',
    iteraciones, f'\nEl valor de π es: {math.pi}',
    f'\nEl valor que obtuvimos es: {piaprox}',
    f'\nCon un error absoluto de: {abs(piaprox - math.pi)}\n')

piaprox, iteraciones = arctan_basica(1e-4)

print('El numero mínimo de términos necesarios para que |4Pn(1)−π| < 10^-4 es:\n',
    iteraciones, f'\nEl valor de π es: {math.pi}',
    f'\nEl valor que obtuvimos es: {piaprox}',
    f'\nCon un error absoluto de: {abs(piaprox - math.pi)}')
```

```python
def arctan(x, tolerancia, maxiter=100000):
    suma=0
    n=0
    for i in range(1, maxiter+1):
        sm= ((-1)**(i+1)) * (x**(2*i-1)) / (2*i-1)
        suma += sm
        n += 1
        if abs(sm) < tolerancia:
            return suma, n

def piaprox_mejorado(tolerancia):
    
    arco1, iter1 = arctan(1/2, tolerancia)
    arco2, iter2 = arctan(1/3, tolerancia)
    piaprox = 4* (arco1 + arco2)
    return piaprox, iter1, iter2

print('<<<<<<<<<<Aproximación de pi con mejora de convergencia>>>>>>>>>>\n')

piaprox, iter1, iter2 = piaprox_mejorado(1e-3)

print('--------------Con un error de 1e-3--------------\n')

print(f'Aproximación de π:  {piaprox}',
    f'\nEl valor real de π es:  {math.pi}',
    f'\nError absoluto de la aproximación:  {abs(piaprox - math.pi)}',
    f'\nTérminos usados en arctan(1/2):  {iter1}', 
    f'\nTérminos usados en arctan(1/3):  {iter2}\n')

print('--------------Con un error de 1e-6--------------\n')

piaprox, iter1, iter2 = piaprox_mejorado(1e-6)

print(f'Aproximación de π:  {piaprox}',
    f'\nEl valor real de π es:  {math.pi}',
    f'\nError absoluto de la aproximación:  {abs(piaprox - math.pi)}',
    f'\nTérminos usados en arctan(1/2):  {iter1}', 
    f'\nTérminos usados en arctan(1/3):  {iter2}')
```

```python
def piaprox_alta_precision(tolerancia):
    arco1, iter1 = arctan(1/5, tolerancia)
    arco2, iter2 = arctan(1/239, tolerancia)
    piaprox = 4* (4*arco1 - arco2)
    return piaprox, iter1, iter2

print('<<<<<<<<<<Aproximación de pi con método de alta precisión>>>>>>>>>>\n')

piaprox, iter1, iter2 = piaprox_alta_precision(1e-3)

print('--------------Con un error de 1e-3--------------\n')

print(f'Aproximación de π:  {piaprox}',
    f'\nEl valor real de π es:  {math.pi}',
    f'\nError absoluto de la aproximación:  {abs(piaprox - math.pi)}',
    f'\nTérminos usados en arctan(1/2):  {iter1}', 
    f'\nTérminos usados en arctan(1/3):  {iter2}\n')

print('--------------Con un error de 1e-6--------------\n')

piaprox, iter1, iter2 = piaprox_alta_precision(1e-6)

print(f'Aproximación de π:  {piaprox}',
    f'\nEl valor real de π es:  {math.pi}',
    f'\nError absoluto de la aproximación:  {abs(piaprox - math.pi)}',
    f'\nTérminos usados en arctan(1/2):  {iter1}', 
    f'\nTérminos usados en arctan(1/3):  {iter2}')
```

Se puede notar que en orden, los metodos mas eficientes son
1. Metodo de alta precisión, solo necesita de maximo 3 iteraciones para converger con una precision de 1e-3, y 5 iteraciones para converger con una precision de 1e-6
2. Metodo de mejora de convergencia, necesita de maximo 5 iteraciones para converger con una precision de 1e-3, y 9 iteraciones para converger con una precision de 1e-6
3. La aproximacion basica, la menos eficiente pues necesita de 1000 iteraciones para converger con una precision de 1e-3, y 10000 iteraciones para converger con una precision de 1e-6

### Cuarto problema

```python
def exponencial(x, tol=1e-7, maxiter=10000):
    suma = 0
    terminos = []
    for n in range(maxiter):
        term = (-1)**n * x**n / math.factorial(n)
        suma += term
        terminos.append(term)
        sigtermino = abs(((-x)**(n+1)) / math.factorial(n+1))
        if sigtermino <= tol * abs(suma):
            return suma, n+1, terminos

print('<<<<<<<<<< Aproximación de e^(-x) usando serie >>>>>>>>>>\n')
lista = np.linspace(0.1, 1.0, 10)
for x in lista:    
    resultado, n, terminos = exponencial(x)

    print(f'Para x = {x}:\n'
        f'El número mínimo de términos N necesarios para que el siguiente término sea menor que 10^-7 veces la suma parcial es:\n{n}\n'
        f'El valor exacto de e^(-{x}) es: {np.exp(-x)}\n'
        f'El valor aproximado que obtuvimos es: {resultado}\n'
        f'Con un error absoluto de: {abs(resultado - np.exp(-x))}\n')

```
