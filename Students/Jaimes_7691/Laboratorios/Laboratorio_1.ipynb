### Primer punto

```python
from itertools import combinations

print('<<<<<<Vamos a categorizar un numero>>>>>>\n')

a=int(input("Introduzca el numero entero que desea categorizar:"))

'''
Divisores propios del número
'''

def div_propios(a):
    div_propios = []
    for i in range(1,a):
        if a%i==0:
            div_propios.append(i)
    return div_propios

'''
Numeros defectivos
'''

def numeros_defectivos(a):
    if sum(div_propios(a)) < a:
        return True
    else:
        return False

''' 
Numeros abundantes
'''

def numeros_abundantes(a):
    if sum(div_propios(a)) > a:
        return True
    else:
        return False

''' 
Numeros semiperfectos
'''

def numeros_semiperfectos(a):
    for r in range(1, len(div_propios(a)) + 1):
        for combo in combinations(div_propios(a), r):
            if sum(combo) == a:
                return True
    return False

''' 
Numeros perfectos
'''

def numeros_perfectos(a):
    if sum(div_propios(a)) == a:
        return True
    else:
        return False

'''
Numeros primos
'''

def numeros_primos(a):
    if len(div_propios(a)) == 1:
        return True
    else:
        return False

print(
    'El numero que se va a categorizar es ', a, '\n',
    '\nNumero defectivo: ', numeros_defectivos(a),
    '\nNumero abundante: ', numeros_abundantes(a),
    '\nNumero semiperfecto: ', numeros_semiperfectos(a),
    '\nNumero perfecto: ', numeros_perfectos(a),
    '\nNumero primo: ', numeros_primos(a)
    )
```

```python
print('<<<<<<<<<<<<<<<<<<<<EJERCICIO EXTRA>>>>>>>>>>>>>>>>>>>>')

primeros=[]
n=0
while len(primeros) < 30:
    n += 1
    if numeros_defectivos(n) == True:
        primeros.append(n)

print('Los primeros 30 numeros defectivos son:\n', primeros, '\n')

primeros=[]
n=0
while len(primeros) < 30:
    n += 1
    if numeros_abundantes(n) == True:
        primeros.append(n)

print('Los primeros 30 numeros abundantes son:\n', primeros, '\n')

primeros=[]
n=0
while len(primeros) < 30:
    n += 1
    if numeros_semiperfectos(n) == True:
        primeros.append(n)

print('Los primeros 30 numeros semiperfectos son:\n', primeros, '\n')

primeros=[]
n=0
while len(primeros) < 4:
    n += 1
    if numeros_perfectos(n) == True:
        primeros.append(n)

print('Los primeros 4 numeros perfectos son (se limita):\n', primeros, '\n')

primeros=[]
n=0
while len(primeros) < 30:
    n += 1
    if numeros_primos(n) == True:
        primeros.append(n)

print('Los primeros 30 numeros primos son:\n', primeros)
```

```python
'''
Números de Catalan
'''

n = 50

def num_catalan(n):
    if n == 0:
        return 1
    else:
        return (4*n-2) / (n+1) * num_catalan(n-1)

print('Su numero de catalan numero ', n, ' es:\n', num_catalan(n), '\n')

'''
Máximo común divisor
'''

m = 108
n = 192

def mcd(m, n):
    if n == 0:
        return m
    else:
        return mcd(n, m%n)

print(f'El máximo común divisor de {m} y {n} es:\n{mcd(m,n)}')
```

### Cuarto punto

```python
e = 10
m = 9.11e-31
v = 9
h = 1.055e-34

def ev2j(x):
    return (1.602e-19)*x
#Cambiamos de eV a Joules
E = ev2j(e)
V = ev2j(v)
#Calculamos los vectores de onda
k1 = np.sqrt(2 * m * E) / h 
k2 = np.sqrt(2 * m * (E-V)) / h
# Probabilidades
T = (4 * k1 * k2) / (k1 + k2)**2
R = ((k1 - k2) / (k1 + k2))**2

print('Las probabilidades para transmisión dado los datos iniciales es de:\n', round(T,4),
    '\nLas probabilidades para reflexión dado los datos iniciales es de:\n', round(R,4))
print(T+R)
```

### Quinto punto

```python
def orbita(l1, v1):

    '''
    Calcula los parametros orbitales de un cuerpo que orbita el sol a partir de
    su distancia al sol en el perihelio y su velocidad en ese punto
    Entrada:
        l1 (float): Distancia al sol en el perihelio en metros
        v1 (float): Velocidad del cuerpo en el perihelio en metros por segundo
    Salida:
        l2 (float): Distancia al sol en el afelio en metros
        v2 (float): Velocidad del cuerpo en el afelio en metros por segundo
        periodo_orbital (float): Periodo orbital del cuerpo
        excentricidad_orbital (float): Excentricidad de la órbita
    '''

    M = 1.9891e30 #Masa del sol constante
    G = 6.6738e-11 #Constante gravitacional 

    a = 1
    b = - ((2 * G * M) / (v1 * l1))
    c = - ((v1**2) - (2 * G * M) / l1)

    raiz1 = (- b + np.sqrt(b**2 - 4 * a * c)) / (2 * a)
    raiz2 = (- b - np.sqrt(b**2 - 4 * a * c)) / (2 * a)

    if raiz1 <= raiz2:
        v2 = raiz1
    else:
        v2 = raiz2
    
    l2 = l1 * v1 / v2

    semieje_mayor = (1 / 2) * (l1 + l2)
    semieje_menor = np.sqrt(l1 * l2)
    periodo_orbital = (2 * np.pi * semieje_mayor * semieje_menor) / (l1 * v1)
    excentricidad_orbital = (l2 - l1) / (l2 + l1)
    return l2, v2, periodo_orbital, excentricidad_orbital
```

```python
print('//////////// Vamos a calcular las propiedades de las órbitas de la tierra ////////////\n')

l1tierra = 1.4710e11
v1tierra = 3.0287e4
l2tierra, v2tierra, Ttierra, etierra = orbita(l1tierra, v1tierra)

print('La distancia al sol en el perihelio en metros ingresada fue: ', l1tierra,
    '\nLa velocidad en el perihelio ingresada en m/s fue: ', v1tierra, '\n')

print('<<<<<<<<<<Resultados>>>>>>>>>>\n', 
    f'\nLa distancia al sol en el afelio (l2):  {l2tierra}',
    f'\nLa velocidad en el afelio (v2):  {v2tierra}',
    f'\nEl periodo orbital (T):  {Ttierra}',
    f'\nEl periodo orbital en años (T):  {Ttierra*3.171e-8}',
    f'\nLa excentricidad orbital (e):  {etierra}')

print("\n======Se confirma el periodo de aproximadamente un año======\n")

print('//////////// Vamos a calcular las propiedades de las órbitas del cometa Halley ////////////\n')

l1halley = 8.7830e10
v1halley = 5.4529e4
l2halley, v2halley, Thalley, ehalley = orbita(l1halley, v1halley)

print('La distancia al sol en el perihelio en metros ingresada fue: ', l1halley,
    '\nLa velocidad en el perihelio ingresada en m/s fue: ', v1halley, '\n')

print('<<<<<<<<<<Resultados>>>>>>>>>>\n', 
    f'\nLa distancia al sol en el afelio (l2):  {l2halley}',
    f'\nLa velocidad en el afelio (v2):  {v2halley}',
    f'\nEl periodo orbital (T):  {Thalley}',
    f'\nEl periodo orbital en años (T):  {Thalley*3.171e-8}',
    f'\nLa excentricidad orbital (e):  {ehalley}')

print("\n======Se confirma el periodo de aproximadamente 76 años======")

```

### Sexto punto

```python
g = 9.81
teta = 45 * (np.pi/180)
vo = 4
vox = np.cos(teta)*vo
voy = np.sin(teta)*vo

def xt(t):
    return vox * t

def yt(t):
    return voy*t - 0.5*g*(t**2)

def vyt(t):
    return voy - g*t

t_total = 2 * voy /g

t = np.linspace(0, t_total, 10000)

plt.style.use('bmh')
plt.title(f'Movimiento parabolico de un proyectil \nv0 = 4.0 m/s, angulo = 45°')
plt.xlabel('Movimiento horizontal (m)')
plt.ylabel('Movimiento vertical (m)')

plt.plot(xt(t), yt(t), 'g--', label = 'Trayectoria', color='purple')
plt.legend()
plt.show()
```

### Septimo punto

```python
def pn(n):
    x = symbols("x")
    return (1 / ((factorial(n)) * (2**n))) * diff((x**2 - 1)**n , x, n)

print("Usando la Fórmula de Rodrigues P0 y P1 son:\n",
    "p0(x) =" , pn(0),
    "\n p1(x) =" , pn(1))
```

```python
def p_recurrencia(x, j):
    if j == 0:
        return 1
    elif j == 1:
        return x
    else:
        return (((2*j-1)*x*p_recurrencia(x, j-1)) - ((j-1)*p_recurrencia(x, j-2))) / (j)

print('Los polinomios de Legendre por recurrencia de P1 a P5')

x=np.linspace(-1,1,1000)

plt.title(r'Polinomios de Legendre $P_n$')
plt.xlabel("x")
plt.ylabel(r"$P_n(x)$")

plt.plot(x , p_recurrencia(x,1) , label=r"$P_1(x)$")
plt.plot(x , p_recurrencia(x,2) , label=r"$P_2(x)$")
plt.plot(x , p_recurrencia(x,3) , label=r"$P_3(x)$")
plt.plot(x , p_recurrencia(x,4) , label=r"$P_4(x)$")
plt.plot(x , p_recurrencia(x,5) , label=r"$P_5(x)$")
```
