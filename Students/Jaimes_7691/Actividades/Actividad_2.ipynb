## Problema 1

Una persona conduce una moto con rapidez constante $v_m$, mientras que un policia lo hace con aceleración constante $a_p$ (iniciando desde el reposo). Suponga que ambos movimientos inician en $t=0$, haga una gráfica para el movimiento de la moto y del policia, donde el usuario ingrese la velocidad (moto) y aceleración (policia). Además determine el instante $t$ en el cual se encuentran ambos. Ambos movimientos iniciarán en $x_i = 0$

Pruebe: para $v_m$ = 15m/s y $a_p$ = 3 m/s², el tiempo deberia ser 10 s


$$x_f = x_i + v_m\cdot t$$

$$x_f = x_i + v_o\cdot t + \frac{1}{2}a_p\cdot t^2$$

```python
a_pol = float(input('Ingrese la aceleracion del policia (m/s^2): '))
v_moto = float(input('Ingresa la velocidad de la moto (m/s): '))

t_encuentro = (2 * v_moto) / a_pol #Se encuentra igualando las ecuaciones
print("El policia se necuentra con el motociclista a los", t_encuentro,"segundos.")

t = np.linspace(0, t_encuentro*1.5, 100)
x_moto = v_moto * t
x_pol = 0.5 * a_pol * t**2

plt.title('Movimiento de la moto y el policia')
plt.xlabel('Tiempo (s)')
plt.ylabel('Posicion (m)')

plt.plot(t, x_moto, label='Moto', color='purple')
plt.plot(t, x_pol, label='Policia', linestyle='--')
plt.plot(t_encuentro, v_moto * t_encuentro, 'ro', label='Encuentro', color='pink')
plt.axvline(x=t_encuentro, linestyle=':', color='gray')
plt.grid(True)
plt.legend(fontsize=12)
plt.show()
```

## Problema 2

El potencial Lennard-Jones es un modelo simple que aproxima la interacción entre un par de moléculas neutras. Este está definido como:

$$
V(r) = 4\epsilon \left[ \left( \frac{\sigma}{r} \right)^{12} - \left( \frac{\sigma}{r} \right)^{6} \right]
$$

Donde:
- $V(r)$ es el potencial en función de la distancia interatómica $r$.
- $\epsilon$ es la profundidad del pozo de potencial, es decir, la energía mínima que describe la interacción entre dos partículas.
- $\sigma$ es la distancia a la cual el potencial es cero, también conocido como el parámetro de distancia.
- $r$ es la distancia entre las dos partículas.

a. Escriba una función para encontrar modelar el potencial LJ.


b. Usando $\sigma = \epsilon = 1$ demuestre que la distancia de equilibrio $r_m$ es aproximadamente $r_m = 2^{1/6}\sigma$. Para esto realice un gráfico del potencial, especificando los ejes. Trunque la serie en $r_c = 2.5\sigma$.


c. Verifique que $r_m = 2^{1/6}\sigma$ se cumple para diferentes valores de $\sigma$, dejando $\epsilon$ constante.


```python
#Problema 2
##a. Definimos la funcion para modelar el potencial de Lennard-Jones
def V_lennad_jones(r):
    ''' 
    Aproxima la interacción entre un par de moleculas usando el potencial de Lennard-Jones.
    Entrada: 
        r: distancia entre las dos particulas
        epsilon:es la profundidad del pozo de potencial, la energia minima que 
                describe la interaccion de dos particulas.
        sigma: distancia a la cual el potencial es cero. 
    Salida:
        V(r): el potencial en funcion de la distancia interatomica r
    '''
    return 4 * epsilon * ((sigma / r)**12 - (sigma / r)**6)

#b. 

sigma = 1
epsilon = 1

rc = (2.5 * sigma) # truncamos el potencial en 2.5 sigma

eje_x = np.linspace(0.1, rc, 100) # rango de distancias
eje_y = V_lennad_jones(eje_x) # Genera el potencial para cada distancia en el rango

plt.style.use('bmh') # Cambiamos el estilo de la grafica
plt.figure(figsize=(24,8)) # Ajustamos el tamaño de la figura
plt.grid(True, which='both', linestyle=':', linewidth=1, color='#010101') 

plt.title('Potencia de Lennard-Jones \n σ = 1       ϵ = 1', fontsize=20) 
plt.xlabel('Distancia entre atomos (r)', fontsize=15)
plt.ylabel('Potencial Lennard-Jones (V)', fontsize=15)

plt.plot(eje_x, eje_y, 'g--', label='Potencial de Lennard-Jones', color= '#c733ff') 
plt.plot(2**(1/6)*sigma, V_lennad_jones(2**(1/6)*sigma), 'ro', label='Minimo del potencial', color= '#d72a79', markersize=10) # Distancia de equilibrio rm

# Ajustamos el rango que se va a graficar
plt.xlim(0, rc) 
plt.ylim(-abs(1.5*V_lennad_jones(2**(1/6)*sigma)), abs(V_lennad_jones(2**(1/6)*sigma)))
plt.legend()
plt.show()


#c. Verificamos que la distancia de quilibrio rm=2**(1/6)*sigma se cumples para diferentes valores de sigma


fig, axs = plt.subplots(3,2, figsize=(15, 10))

plt.subplots_adjust(hspace=0.5, wspace=0.5)

sigma = 20

eje_x = np.linspace(0.1, 2.5*sigma, 100)
eje_y = V_lennad_jones(eje_x)

axs[0,0].grid(True, linestyle=':', linewidth=1)
axs[0,0].set_title('Potencia de Lennard-Jones \n σ = 20       ϵ = 1', fontsize=12)
axs[0,0].set_xlabel('Distancia entre atomos (r)', fontsize=10)
axs[0,0].set_ylabel('Potencial Lennard-Jones (V)', fontsize=10)

axs[0,0].plot(eje_x, eje_y, 'g--', label='Potencial de Lennard-Jones')
axs[0,0].plot(2**(1/6) * sigma, V_lennad_jones(2**(1/6) * sigma), 'ro', label='Minimo del potencial', markersize=6)
axs[0,0].set_xlim(0, 2.5 * sigma)
axs[0,0].set_ylim(-abs(1.5*V_lennad_jones(2**(1/6) * sigma)), abs(V_lennad_jones(2**(1/6) * sigma)))

axs[0,0].legend(fontsize=6)


sigma = 60

eje_x = np.linspace(0.1, 2.5*sigma, 100)
eje_y = V_lennad_jones(eje_x)

axs[0,1].grid(True, linestyle=':', linewidth=1)
axs[0,1].set_title('Potencia de Lennard-Jones \n σ = 60       ϵ = 1', fontsize=12)
axs[0,1].set_xlabel('Distancia entre atomos (r)', fontsize=10)
axs[0,1].set_ylabel('Potencial Lennard-Jones (V)', fontsize=10)

axs[0,1].plot(eje_x, eje_y, 'g--', label='Potencial de Lennard-Jones')
axs[0,1].plot(2**(1/6) * sigma, V_lennad_jones(2**(1/6) * sigma), 'ro', label='Minimo del potencial', markersize=6)
axs[0,1].set_xlim(0, 2.5 * sigma)
axs[0,1].set_ylim(-abs(1.5*V_lennad_jones(2**(1/6) * sigma)), abs(V_lennad_jones(2**(1/6) * sigma)))

axs[0,1].legend(fontsize=6)

sigma = 138

eje_x = np.linspace(0.1, 2.5*sigma, 100)
eje_y = V_lennad_jones(eje_x)

axs[1,0].grid(True, linestyle=':', linewidth=1)
axs[1,0].set_title('Potencia de Lennard-Jones \n σ = 138       ϵ = 1', fontsize=12)
axs[1,0].set_xlabel('Distancia entre atomos (r)', fontsize=10)
axs[1,0].set_ylabel('Potencial Lennard-Jones (V)', fontsize=10)

axs[1,0].plot(eje_x, eje_y, 'g--', label='Potencial de Lennard-Jones')
axs[1,0].plot(2**(1/6) * sigma, V_lennad_jones(2**(1/6) * sigma), 'ro', label='Minimo del potencial', markersize=6)
axs[1,0].set_xlim(0, 2.5 * sigma)
axs[1,0].set_ylim(-abs(1.5*V_lennad_jones(2**(1/6) * sigma)), abs(V_lennad_jones(2**(1/6) * sigma)))

axs[1,0].legend(fontsize=6)

sigma = 187

eje_x = np.linspace(0.1, 2.5*sigma, 100)
eje_y = V_lennad_jones(eje_x)

axs[1,1].grid(True, linestyle=':', linewidth=1)
axs[1,1].grid(True, linestyle=':', linewidth=1)
axs[1,1].set_title('Potencia de Lennard-Jones \n σ = 1000       ϵ = 1', fontsize=12)
axs[1,1].set_xlabel('Distancia entre atomos (r)', fontsize=10)
axs[1,1].set_ylabel('Potencial Lennard-Jones (V)', fontsize=10)


axs[1,1].plot(eje_x, eje_y, 'g--', label='Potencial de Lennard-Jones')
axs[1,1].plot(2**(1/6) * sigma, V_lennad_jones(2**(1/6) * sigma), 'ro', label='Minimo del potencial', markersize=6)
axs[1,1].set_xlim(0, 2.5 * sigma)
axs[1,1].set_ylim(-abs(1.5*V_lennad_jones(2**(1/6) * sigma)), abs(V_lennad_jones(2**(1/6) * sigma)))

axs[1,1].legend(fontsize=6)

sigma = 1000

eje_x = np.linspace(0.1, 2.5*sigma, 100)
eje_y = V_lennad_jones(eje_x)

axs[2,0].grid(True, linestyle=':', linewidth=1)
axs[2,0].set_title('Potencia de Lennard-Jones \n σ = 1000       ϵ = 1', fontsize=12)
axs[2,0].set_xlabel('Distancia entre atomos (r)', fontsize=10)
axs[2,0].set_ylabel('Potencial Lennard-Jones (V)', fontsize=10)

axs[2,0].plot(eje_x, eje_y, 'g--', label='Potencial de Lennard-Jones')
axs[2,0].plot(2**(1/6) * sigma, V_lennad_jones(2**(1/6) * sigma), 'ro', label='Minimo del potencial', markersize=6)
axs[2,0].set_xlim(0, 2.5 * sigma)
axs[2,0].set_ylim(-abs(1.5*V_lennad_jones(2**(1/6) * sigma)), abs(V_lennad_jones(2**(1/6) * sigma)))
axs[2,0].grid(True, which='both', linestyle=':', linewidth=1)

axs[2,0].legend(fontsize=6)

sigma = 90000

axs[2,1].grid(True, linestyle=':', linewidth=1)
axs[2,1].grid(True, linestyle=':', linewidth=1)
axs[2,1].set_title('Potencia de Lennard-Jones \n σ = 90000       ϵ = 1', fontsize=12)
axs[2,1].set_xlabel('Distancia entre atomos (r)', fontsize=10)
axs[2,1].set_ylabel('Potencial Lennard-Jones (V)', fontsize=10)

eje_x = np.linspace(0.1, 2.5*sigma, 100)
eje_y = V_lennad_jones(eje_x)

axs[2,1].plot(eje_x, eje_y, 'g--', label='Potencial de Lennard-Jones')
axs[2,1].plot(2**(1/6) * sigma, V_lennad_jones(2**(1/6) * sigma), 'ro', label='Minimo del potencial', markersize=6)
axs[2,1].set_xlim(0, 2.5 * sigma)
axs[2,1].set_ylim(-abs(1.5*V_lennad_jones(2**(1/6) * sigma)), abs(V_lennad_jones(2**(1/6) * sigma)))
axs[2,1].grid(True, which='both', linestyle=':', linewidth=1)

axs[2,1].legend(fontsize=6)
```
